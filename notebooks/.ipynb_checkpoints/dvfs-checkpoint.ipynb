```python
import numpy as np
import matplotlib.pyplot as plt

# Benchmarks / display names
bench_order = ["perl", "gcc", "mcf", "omnetpp", "xalan", "dsj", "leela", "xz", "exch", "foto"]
xtick_labels = ["mean", "perl", "gcc", "mcf", "omnetpp", "xalan", "dsj", "leela", "xz", "exch", "foto"]
configs = ["SRAM7", "MRAM14", "M+DVFS"]

# Hardcoded decomposition data: [E_nc_dyn, E_nc_static, E_llc_dyn, E_llc_leak]
raw = {
    "SRAM7": {
        "perl":    [9.7401, 3.8371, 0.0006, 0.1911],
        "gcc":     [15.0535, 6.3904, 0.0002, 0.3183],
        "mcf":     [12.3899, 5.2145, 0.0308, 0.2597],
        "omnetpp": [12.9915, 5.3610, 0.0005, 0.2670],
        "xalan":   [11.8884, 5.0844, 0.0076, 0.2533],
        "dsj":     [10.3180, 4.1263, 0.0001, 0.2055],
        "leela":   [8.6556,  3.3308, 0.0000, 0.1659],
        "xz":      [9.2901,  3.8439, 0.0001, 0.1915],
        "exch":    [8.4176,  3.2238, 0.0000, 0.1606],
        "foto":    [9.6520,  3.8047, 0.0000, 0.1895],
    },
    "MRAM14": {
        "perl":    [9.6505, 3.8018, 0.0006, 0.0388],
        "gcc":     [15.0192, 6.3759, 0.0002, 0.0650],
        "mcf":     [11.6899, 4.9199, 0.0290, 0.0502],
        "omnetpp": [12.9647, 5.3499, 0.0005, 0.0546],
        "xalan":   [11.1989, 4.7895, 0.0072, 0.0488],
        "dsj":     [10.3031, 4.1203, 0.0001, 0.0420],
        "leela":   [8.6550,  3.3306, 0.0000, 0.0340],
        "xz":      [9.2790,  3.8393, 0.0001, 0.0392],
        "exch":    [8.4176,  3.2238, 0.0000, 0.0329],
        "foto":    [9.6518,  3.8046, 0.0000, 0.0388],
    },
    "M+DVFS": {
        "perl":    [9.4635, 3.7281, 0.0006, 0.0380],
        "gcc":     [14.8318, 6.2963, 0.0002, 0.0642],
        "mcf":     [11.6341, 4.8964, 0.0290, 0.0499],
        "omnetpp": [12.7309, 5.2534, 0.0005, 0.0536],
        "xalan":   [11.1887, 4.7852, 0.0072, 0.0488],
        "dsj":     [10.1904, 4.0753, 0.0001, 0.0416],
        "leela":   [8.4588,  3.2551, 0.0000, 0.0332],
        "xz":      [9.2707,  3.8358, 0.0001, 0.0391],
        "exch":    [8.3029,  3.1798, 0.0000, 0.0324],
        "foto":    [9.5276,  3.7556, 0.0000, 0.0383],
    },
}

# Add mean rows
for cfg in configs:
    arr = np.array([raw[cfg][b] for b in bench_order], dtype=float)
    raw[cfg] = {"mean": arr.mean(axis=0), **raw[cfg]}

plot_order = ["mean"] + bench_order
vals = {cfg: np.array([raw[cfg][b] for b in plot_order], dtype=float) for cfg in configs}

# Normalize each bar by its own total / own LLC total
norm_total = {}
norm_llc = {}
for cfg in configs:
    arr = vals[cfg]
    total = arr.sum(axis=1, keepdims=True)
    llc_total = (arr[:, 2] + arr[:, 3]).reshape(-1, 1)

    norm_total[cfg] = arr / total

    llc_norm = np.zeros_like(arr)
    llc_norm[:, 2] = np.divide(arr[:, 2], llc_total[:, 0], out=np.zeros(len(plot_order)), where=llc_total[:, 0] > 0)
    llc_norm[:, 3] = np.divide(arr[:, 3], llc_total[:, 0], out=np.zeros(len(plot_order)), where=llc_total[:, 0] > 0)
    norm_llc[cfg] = llc_norm

# Plot setup
x = np.arange(len(plot_order))
w = 0.22
offsets = [-w, 0, w]

c_nc_dyn = "#4C72B0"
c_nc_static = "#C9D3DD"
c_llc_dyn = "#D8A03A"
c_llc_leak = "#D9534F"

fig, axes = plt.subplots(
    2, 1, figsize=(13, 8.8), sharex=True,
    gridspec_kw={"hspace": 0.18}
)

# --- Top panel ---
ax = axes[0]
for i, cfg in enumerate(configs):
    xpos = x + offsets[i]
    arr = norm_total[cfg]

    ax.bar(xpos, arr[:, 0], width=w, color=c_nc_dyn, edgecolor="gray", linewidth=0.8,
           label="Non-cache dynamic" if i == 0 else None)
    ax.bar(xpos, arr[:, 1], width=w, bottom=arr[:, 0], color=c_nc_static, edgecolor="gray", linewidth=0.8,
           label="Non-cache static" if i == 0 else None)
    ax.bar(xpos, arr[:, 2], width=w, bottom=arr[:, 0] + arr[:, 1], color=c_llc_dyn, edgecolor="gray", linewidth=0.8,
           label="LLC dynamic" if i == 0 else None)
    ax.bar(xpos, arr[:, 3], width=w, bottom=arr[:, 0] + arr[:, 1] + arr[:, 2], color=c_llc_leak, edgecolor="gray", linewidth=0.8,
           label="LLC leakage" if i == 0 else None)

ax.set_ylabel("Normalized Total System Energy")
ax.set_ylim(0, 1.12)
ax.grid(axis="y", alpha=0.25)
ax.set_axisbelow(True)
ax.legend(ncol=4, loc="upper center", bbox_to_anchor=(0.5, 1.08), frameon=False, fontsize=9)

ax.text(
    0.02, 0.05, "Total system energy",
    transform=ax.transAxes, fontsize=10,
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.9, pad=2.5)
)

# --- Bottom panel ---
ax2 = axes[1]
for i, cfg in enumerate(configs):
    xpos = x + offsets[i]
    arr = norm_llc[cfg]

    ax2.bar(xpos, arr[:, 2], width=w, color=c_llc_dyn, edgecolor="gray", linewidth=0.8,
            label="LLC dynamic" if i == 0 else None)
    ax2.bar(xpos, arr[:, 3], width=w, bottom=arr[:, 2], color=c_llc_leak, edgecolor="gray", linewidth=0.8,
            label="LLC leakage" if i == 0 else None)

ax2.set_ylabel("Normalized LLC Energy")
ax2.set_ylim(0, 1.12)
ax2.grid(axis="y", alpha=0.25)
ax2.set_axisbelow(True)

ax2.text(
    0.02, 0.05, "L3-only energy",
    transform=ax2.transAxes, fontsize=10,
    bbox=dict(facecolor="white", edgecolor="none", alpha=0.9, pad=2.5)
)

# X ticks
ax2.set_xticks(x)
ax2.set_xticklabels(xtick_labels)

# Per-bar config labels
for i, cfg in enumerate(configs):
    for j in range(len(plot_order)):
        ax2.text(
            x[j] + offsets[i], -0.045, cfg,
            ha="center", va="top", rotation=90, fontsize=7, clip_on=False
        )

# Move xlabel farther down so it does not overlap the rotated tick text
ax2.set_xlabel("Within each workload cluster (left-right): SRAM7, MRAM14, M+DVFS", labelpad=28)

# Extra bottom margin for rotated labels + xlabel
plt.subplots_adjust(bottom=0.16)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# Global style
# -----------------------------
plt.rcParams.update({
    "font.size": 11,
    "axes.labelsize": 11,
    "axes.titlesize": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})

# =========================================================
# Figure 1: Single-Core Speedup Decomposition
# Data source: Table 3 geomeans
# =========================================================

capacities = ["16 MB", "32 MB", "128 MB"]

# Geomeans from Table 3
spd_mram = np.array([1.011, 1.011, 1.012])   # Spd_MRAM
spd_dvfs = np.array([1.018, 1.025, 1.076])   # Spd_DVFS

# Decompose total speedup over SRAM14 baseline = 1.0
cache_gain = spd_mram - 1.0
dvfs_gain  = spd_dvfs - spd_mram

fig, ax = plt.subplots(figsize=(7.2, 3.2), dpi=150)

x = np.arange(len(capacities))
w = 0.62

cache_color = "#4C78A8"   # blue
dvfs_color  = "#F58518"   # orange

ax.bar(
    x, cache_gain, width=w, bottom=1.0,
    color=cache_color, edgecolor="black", linewidth=1.0,
    label="Cache-level gain (SRAM14 → MRAM)"
)
ax.bar(
    x, dvfs_gain, width=w, bottom=1.0 + cache_gain,
    color=dvfs_color, edgecolor="black", linewidth=1.0,
    label="Additional DVFS gain (MRAM → MRAM+DVFS)"
)

# Keep only stack labels inside the bars
for xi, cg, dg in zip(x, cache_gain, dvfs_gain):
    ax.text(
        xi, 1.0 + cg/2, f"+{cg*100:.1f}%",
        ha="center", va="center", color="white", fontweight="bold"
    )
    ax.text(
        xi, 1.0 + cg + dg/2, f"+{dg*100:.1f}%",
        ha="center", va="center", color="black", fontweight="bold"
    )

ax.set_xticks(x)
ax.set_xticklabels(capacities)
ax.set_ylabel("Normalized Speedup\n(Baseline = SRAM14 = 1.0×)")
ax.set_xlabel("LLC Capacity")
ax.set_title("Single-Core Speedup Decomposition")
ax.set_ylim(1.0, 1.10)
ax.set_yticks(np.arange(1.00, 1.101, 0.02))
ax.grid(axis="y", linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(frameon=False, ncol=1, loc="upper left")

plt.tight_layout()
plt.show()


# =========================================================
# Figure 2: Multicore Enabling Margin
# Data source: Table 7 geomeans
# =========================================================

groups = ["n=4\n16 MB", "n=4\n32 MB", "n=4\n128 MB",
          "n=8\n16 MB", "n=8\n32 MB", "n=8\n128 MB"]

# Geomeans from Table 7
spd_dvfs_mc = np.array([1.120, 1.162, 1.049, 1.029, 1.066, 1.039])
spd_cf_mc   = np.array([1.049, 1.049, 1.064, 1.028, 1.017, 1.049])

gap_pp = (spd_dvfs_mc - spd_cf_mc) * 100.0

fig, ax = plt.subplots(figsize=(7.2, 3.6), dpi=150)

x = np.arange(len(groups))
w = 0.34

dvfs_color = "#54A24B"   # green
cf_color   = "#E45756"   # red

bars1 = ax.bar(
    x - w/2, spd_dvfs_mc, width=w,
    color=dvfs_color, edgecolor="black", linewidth=1.0,
    label="MRAM+DVFS"
)
bars2 = ax.bar(
    x + w/2, spd_cf_mc, width=w,
    color=cf_color, edgecolor="black", linewidth=1.0,
    label="MRAM+DVFS (using SRAM leakage)"
)

# Baseline line
ax.axhline(1.0, color="black", linewidth=1.0)

# Keep dashed gap connectors, but remove text annotations
for i, (y1, y2, g) in enumerate(zip(spd_dvfs_mc, spd_cf_mc, gap_pp)):
    y_low, y_high = sorted([y1, y2])
    ax.plot([x[i], x[i]], [y_low, y_high], linestyle="--", color="gray", linewidth=1.2)

# Visual separator between 4-core and 8-core groups
ax.axvline(2.5, color="gray", linestyle=":", linewidth=1.2)

ax.set_xticks(x)
ax.set_xticklabels(groups)
ax.set_ylabel("Per-core Average Speedup\n(Baseline = SRAM14 = 1.0×)")
ax.set_xlabel("Multicore Configuration")
ax.set_title("Multicore Enabling Margin")
ax.set_ylim(1.0, 1.19)
ax.set_yticks(np.arange(1.00, 1.191, 0.03))
ax.grid(axis="y", linestyle="--", alpha=0.4)
ax.set_axisbelow(True)
ax.legend(frameon=False, loc="upper left")

plt.tight_layout()
plt.show()
```
