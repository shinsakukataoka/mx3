```python
import math
import pandas as pd

# --------------------------------------------------
# Config
# --------------------------------------------------
CSV_PATH = "/home/skataoka26/COSC_498/devices/results/cache_sweep/uniform_370/nvsim_uniform_370.csv"
FREQ_GHZ = 2.2
CAPS_TO_SHOW = [16, 32, 128]

# --------------------------------------------------
# Helpers
# --------------------------------------------------
def ns_to_cycles(ns, freq_ghz=FREQ_GHZ):
    return math.ceil(ns * freq_ghz)

def infer_tech(path):
    s = str(path)
    if "/SRAM/" in s:
        return "SRAM"
    if "/MRAM/" in s:
        return "MRAM"
    return "UNKNOWN"

# --------------------------------------------------
# Load
# --------------------------------------------------
df = pd.read_csv(CSV_PATH)
df["technology"] = df["stdout_path"].apply(infer_tech)

# keep only the capacities you want
df = df[df["cap_mb"].isin(CAPS_TO_SHOW)].copy()

# --------------------------------------------------
# Convert fields for display
# --------------------------------------------------
df["read_cycles"] = df["lat_hit_ns"].apply(ns_to_cycles)
df["write_cycles"] = df["lat_write_ns"].apply(ns_to_cycles)

# --------------------------------------------------
# Pick one row per (capacity, tech, node)
# If multiple optimization-target rows exist, choose the one with
# minimum effective latency at f_w = 0.7 just to make the table unique.
# --------------------------------------------------
F_W = 0.7
df["L_eff"] = (1 - F_W) * df["lat_hit_ns"] + F_W * df["lat_write_ns"]

idx = df.groupby(["cap_mb", "technology", "node_nm"])["L_eff"].idxmin()
show = df.loc[idx].copy()

show = show.sort_values(
    by=["cap_mb", "technology", "node_nm"],
    key=lambda s: s.map({"SRAM": 0, "MRAM": 1}) if s.name == "technology" else s
)

# --------------------------------------------------
# Print like the example
# --------------------------------------------------
header = (
    f"{'MB':>4}  "
    f"{'Tech':<6}  "
    f"{'Node':<8}  "
    f"{'Read Lat.':>10}  "
    f"{'Write Lat.':>11}  "
    f"{'Leak.':>8}  "
    f"{'Area':>8}"
)

subheader = (
    f"{'':>4}  "
    f"{'':<6}  "
    f"{'(nm)':<8}  "
    f"{'(cycles)':>10}  "
    f"{'(cycles)':>11}  "
    f"{'(mW)':>8}  "
    f"{'(mm^2)':>8}"
)

print(header)
print(subheader)
print("-" * len(header))

prev_cap = None
for _, r in show.iterrows():
    cap = int(r["cap_mb"])
    if prev_cap is not None and cap != prev_cap:
        print()

    print(
        f"{cap:>4}  "
        f"{r['technology']:<6}  "
        f"{str(int(r['node_nm'])) + ' nm':<8}  "
        f"{int(r['read_cycles']):>10}  "
        f"{int(r['write_cycles']):>11}  "
        f"{float(r['leakage_mW']):>8.1f}  "
        f"{float(r['area_total_mm2']):>8.2f}"
    )
    prev_cap = cap
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.lines import Line2D

# --------------------------------------------------
# Config
# --------------------------------------------------
CSV_PATH = "/home/skataoka26/COSC_498/devices/results/cache_sweep/uniform_370/nvsim_uniform_370.csv"
F_W = 0.7
NODE_NM = 14
CAPS_TO_SHOW = [16, 32, 128]

# --------------------------------------------------
# Load
# --------------------------------------------------
df = pd.read_csv(CSV_PATH)

df = df[
    (df["tech"] == "MRAM") &
    (df["node_nm"] == NODE_NM) &
    (df["cap_mb"].isin(CAPS_TO_SHOW))
].copy()

# --------------------------------------------------
# Effective latency / energy
# --------------------------------------------------
df["L_eff"] = (1 - F_W) * df["lat_hit_ns"] + F_W * df["lat_write_ns"]
df["E_eff"] = (1 - F_W) * df["dyn_energy_hit_nJ"] + F_W * df["dyn_energy_write_nJ"]

# --------------------------------------------------
# Plot
# color = optimization target
# marker = capacity
# --------------------------------------------------
fig, ax = plt.subplots(figsize=(4.5, 3.5))

cap_markers = {
    16: "o",
    32: "s",
    128: "^",
}

opt_targets = ["ReadEDP", "ReadLatency", "WriteLatency", "WriteEDP"]

opt_colors = {
    "ReadEDP": "tab:green",
    "ReadLatency": "tab:blue",
    "WriteLatency": "tab:orange",
    "WriteEDP": "tab:red",
}
for _, r in df.iterrows():
    ax.scatter(
        r["L_eff"],
        r["E_eff"],
        marker=cap_markers[int(r["cap_mb"])],
        color=opt_colors[r["opt_target"]],
        s=70,
        alpha=0.85,
    )

ax.set_xlabel("Effective latency (ns)")
ax.set_ylabel("Effective energy (nJ)")
ax.grid(True, alpha=0.3)

# legends
color_handles = [
    Line2D([0], [0], marker='o', linestyle='', color=opt_colors[opt], label=opt, markersize=7)
    for opt in opt_targets[:4]
]

shape_handles = [
    Line2D([0], [0], marker=cap_markers[cap], linestyle='', color='black', label=f"{cap}MB", markersize=7)
    for cap in CAPS_TO_SHOW
]

leg1 = ax.legend(handles=color_handles, title="Optimization target", fontsize=8, title_fontsize=8, loc="best")
ax.add_artist(leg1)
ax.legend(
    handles=shape_handles,
    title="Capacity",
    fontsize=8,
    title_fontsize=8,
    loc="lower right",
    bbox_to_anchor=(1.0, 0.20),
)
plt.tight_layout()
plt.show()
```

```python
#!/usr/bin/env python3
import pandas as pd

CSV = "/home/skataoka26/COSC_498/miniMXE/repro/agg/workload.csv"

NAME_MAP = {
    "505.mcf_r": "mcf",
    "523.xalancbmk_r": "xalancbmk",
    "500.perlbench_r": "perlbench",
    "520.omnetpp_r": "omnetpp",
    "502.gcc_r": "gcc",
    "557.xz_r": "xz",
    "531.deepsjeng_r": "deepsjeng",
    "541.leela_r": "leela",
    "649.fotonik3d_s": "fotonik3d",
    "648.exchange2_s": "exchange2",
}

T4_ORDER = [
    "mcf", "xalancbmk", "perlbench", "omnetpp", "gcc",
    "xz", "deepsjeng", "leela", "fotonik3d", "exchange2"
]

T5_ORDER = ["mcf", "xalancbmk", "deepsjeng", "leela"]

df = pd.read_csv(CSV)

df["short"] = df["workload"].map(NAME_MAP)

# -----------------------------
# Table 4: 16 MB all-MRAM
# -----------------------------
t4 = df[
    (df["n_tag"] == "n1") &
    (df["size_mb"] == 16) &
    (df["config"] == "MRAM14") &
    (df["short"].notna())
].copy()

t4["Write % of Hits"] = (
    t4["l3_write_hits"] / (t4["l3_read_hits"] + t4["l3_write_hits"])
).fillna(0)

t4["Workload"] = pd.Categorical(t4["short"], categories=T4_ORDER, ordered=True)
t4 = t4.sort_values("Workload")

t4_out = pd.DataFrame({
    "Workload": t4["short"],
    "inst/ns": t4["throughput_inst_ns"].map(lambda x: f"{x:.2f}"),
    "L3 Read Hits": t4["l3_read_hits"].map(lambda x: f"{int(round(x)):,}"),
    "L3 Write Hits": t4["l3_write_hits"].map(lambda x: f"{int(round(x)):,}"),
    "Miss Rate": t4["l3_miss_rate"].map(lambda x: f"{100*x:.1f}%"),
    "Write % of Hits": t4["Write % of Hits"].map(lambda x: f"{100*x:.1f}%"),
})

print("\nTable 4: Measured workload behavior at the 16 MB all-MRAM design point\n")
print(t4_out.to_string(index=False))

# -----------------------------
# Table 5: 16MB SRAM7 vs 128MB MRAM14
# -----------------------------
base = df[
    (df["n_tag"] == "n1") &
    (df["size_mb"] == 16) &
    (df["config"] == "SRAM7") &
    (df["short"].isin(T5_ORDER))
][["short", "makespan_s", "long_lat_frac", "mem_cpi_frac"]].copy()

base = base.rename(columns={
    "makespan_s": "Runtime_16MB",
    "long_lat_frac": "LongLat_16MB",
    "mem_cpi_frac": "MemCPI_16MB",
})

target = df[
    (df["n_tag"] == "n1") &
    (df["size_mb"] == 128) &
    (df["config"] == "MRAM14") &
    (df["short"].isin(T5_ORDER))
][["short", "makespan_s", "long_lat_frac", "mem_cpi_frac"]].copy()

target = target.rename(columns={
    "makespan_s": "Runtime_128MB",
    "long_lat_frac": "LongLat_128MB",
    "mem_cpi_frac": "MemCPI_128MB",
})

t5 = base.merge(target, on="short", how="inner")
t5["Workload"] = pd.Categorical(t5["short"], categories=T5_ORDER, ordered=True)
t5 = t5.sort_values("Workload")

t5_out = pd.DataFrame({
    "Workload": t5["short"],
    "Runtime (ms) 16MB": t5["Runtime_16MB"].map(lambda x: f"{x*1000:.1f}"),
    "Runtime (ms) 128MB": t5["Runtime_128MB"].map(lambda x: f"{x*1000:.1f}"),
    "Long-lat Stall 16MB": t5["LongLat_16MB"].map(lambda x: f"{100*x:.1f}%"),
    "Long-lat Stall 128MB": t5["LongLat_128MB"].map(lambda x: f"{100*x:.1f}%"),
    "Mem CPI Frac 16MB": t5["MemCPI_16MB"].map(lambda x: f"{100*x:.1f}%"),
    "Mem CPI Frac 128MB": t5["MemCPI_128MB"].map(lambda x: f"{100*x:.1f}%"),
})

print("\nTable 5: Representative workload dependence when moving from 16 MB SRAM7 to 128 MB MRAM14\n")
print(t5_out.to_string(index=False))
```

# Capacity scaling and HCA

```python
import os, re, math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.lines import Line2D
from pathlib import Path
import yaml

BASE_DIR = Path("~/COSC_498/miniMXE").expanduser()
ALL_CSV    = Path("~/COSC_498/miniMXE/repro/agg/hca.csv").expanduser()

ASSOC = 16

DEVICE_DIR = BASE_DIR / "mx3" / "config" / "devices" / "sunnycove"

def load_nvsim_params(device_dir):
    def _read_yaml(fname):
        with open(device_dir / fname, "r") as f:
            raw = yaml.safe_load(f) or {}
        out = {}
        for size_mb, vals in raw.items():
            out[str(size_mb)] = {
                "r_pj": float(vals.get("r_pj", 0.0)),
                "w_pj": float(vals.get("w_pj", 0.0)),
                "leak_mw": float(vals.get("leak_mw", 0.0)),
            }
        return out

    sram7  = _read_yaml("sram7.yaml")
    sram14 = _read_yaml("sram14.yaml")
    sram32 = _read_yaml("sram32.yaml")
    mram14 = _read_yaml("mram14.yaml")
    mram32 = _read_yaml("mram32.yaml")

    params = {}

    # Pure SRAM techs
    params["sram7"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram7.items()
    }

    params["sram14"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram14.items()
    }

    params["sram32"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram32.items()
    }

    # Pure MRAM techs, but keep SRAM side populated to match old run.yaml logic
    params["mram14"] = {}
    for sz in sorted(set(sram14.keys()) | set(mram14.keys()), key=lambda x: int(x)):
        sv = sram14.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        mv = mram14.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        params["mram14"][sz] = {
            "SRAM_R_pJ": sv["r_pj"],
            "SRAM_W_pJ": sv["w_pj"],
            "SRAM_LEAK_mW": sv["leak_mw"],
            "MRAM_R_pJ": mv["r_pj"],
            "MRAM_W_pJ": mv["w_pj"],
            "MRAM_LEAK_mW": mv["leak_mw"],
        }

    params["mram32"] = {}
    for sz in sorted(set(sram32.keys()) | set(mram32.keys()), key=lambda x: int(x)):
        sv = sram32.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        mv = mram32.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        params["mram32"][sz] = {
            "SRAM_R_pJ": sv["r_pj"],
            "SRAM_W_pJ": sv["w_pj"],
            "SRAM_LEAK_mW": sv["leak_mw"],
            "MRAM_R_pJ": mv["r_pj"],
            "MRAM_W_pJ": mv["w_pj"],
            "MRAM_LEAK_mW": mv["leak_mw"],
        }

    # Explicit hybrid key used by your HCA rows
    params["sram14_mram14"] = dict(params["mram14"])

    return params

NVSIM_PARAMS = load_nvsim_params(DEVICE_DIR)

# (display_name, tech_value, variant_substring)
DESIGNS = [
    ("SRAM_7nm",  "sram7",  "baseline_sram_only"),
    ("SRAM_16nm", "sram14", "baseline_sram_only"),
    ("SRAM_32nm", "sram32", "baseline_sram_only"),
    ("MRAM_16nm", "mram14", "baseline_mram_only"),
    ("MRAM_32nm", "mram32", "baseline_mram_only"),
]

_STATIC_SCENARIOS = [
    ("All-MRAM 16nm",   "mram14", "baseline_mram_only"),
    ("All-SRAM 16nm",   "sram14", "baseline_sram_only"),
    ("Static 4 SRAM",   "sram14_mram14", "noparity_s4_fillmram_sram14_mram14"),
    ("Static 8 SRAM",   "sram14_mram14", "noparity_s8_fillmram_sram14_mram14"),
    ("Static 12 SRAM",  "sram14_mram14", "noparity_s12_fillmram_sram14_mram14"),
]

_COMBINED_COLORS = {
    "All-MRAM 16nm":       "#B22222",
    "All-SRAM 16nm":       "#999999",
    "Static 4 SRAM":       "#1f77b4",
    "Static 8 SRAM":       "#2ca02c",
    "Static 12 SRAM":      "#17becf",
}

_BENCH_MARKERS = ['o', 'v', '^', '<', '>', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X']

plt.rcParams.update({
    "font.size": 12, "axes.titlesize": 13, "axes.labelsize": 12,
    "xtick.labelsize": 9, "ytick.labelsize": 10,
    "legend.fontsize": 10, "figure.dpi": 150,
    "savefig.dpi": 150, "savefig.bbox": "tight",
})

# =============================================================================
# Helpers
# =============================================================================
def gmean(xs):
    xs = np.asarray(list(xs), dtype=float)
    xs = xs[np.isfinite(xs) & (xs > 0)]
    return float(np.exp(np.mean(np.log(xs)))) if xs.size else float("nan")

def _short_bench(name):
    parts = name.split("+")
    shortened = []
    for p in parts:
        s = re.sub(r'^\d+_', '', p)
        s = re.sub(r'_(r|s)$', '', s)
        shortened.append(s)
    return "+".join(shortened)

def compute_w_sram(df_in):
    v = df_in["variant"].fillna("").astype(str)
    w = np.full(len(df_in), np.nan, dtype=float)
    w[v.str.contains("baseline_sram_only", na=False).values] = 1.0
    w[v.str.contains("baseline_mram_only", na=False).values] = 0.0
    sw = pd.to_numeric(df_in.get("sram_ways", np.nan), errors="coerce").values
    ft = df_in.get("fill_to", "").fillna("").astype(str).str.lower().values
    for i in range(len(w)):
        if np.isfinite(w[i]): continue
        try: frac = float(sw[i]) / ASSOC
        except: frac = 0.5
        if ft[i] == "sram": frac = max(frac, 0.6)
        if ft[i] == "mram": frac = min(frac, 0.4)
        w[i] = np.clip(frac, 0.0, 1.0)
    return w

# =============================================================================
# Data Loading
# =============================================================================
def load_all():
    print(f"Reading {ALL_CSV}...")
    df = pd.read_csv(ALL_CSV)
    print(f"Read {len(df)} rows.")

    df.rename(columns={
        "workload": "bench", "device": "tech", "instructions": "insn",
        "l3_rh_sram": "read_hits_sram", "l3_rh_mram": "read_hits_mram",
        "l3_wh_sram": "write_hits_sram", "l3_wh_mram": "write_hits_mram",
    }, inplace=True)

    df["size_mb"] = (
        df["capacity"].astype(str)
        .str.replace(r"^sz", "", regex=True)
        .str.replace(r"M$", "", regex=True)
    )
    df["size_mb"] = pd.to_numeric(df["size_mb"], errors="coerce")
    df["roi_time_s"] = df["elapsed_ns"] / 1e9

    def _build_rw_tag(row):
        ls = row.get("lat_scale")
        dev = str(row.get("tech", ""))
        if pd.isna(ls) or str(ls).strip() == "":
            m_dev = re.search(r"_rd(\d+)x", dev)
            if m_dev: return f"r{m_dev.group(1)}x"
            return ""
        m = re.search(r"(r\d+x_w\d+x)", dev)
        if m: return m.group(1)
        m2 = re.search(r"rd(\d+)x", dev)
        if m2: return f"r{m2.group(1)}x"
        return str(ls)
    df["rw_tag"] = df.apply(_build_rw_tag, axis=1)

    df["tech"] = (df["tech"].astype(str)
                  .str.replace(r"_r\d+x_w\d+x$", "", regex=True)
                  .str.replace(r"_rd\d+x$", "", regex=True))

    def _synth_variant(row):
        hca  = str(row.get("hca", "")).strip().lower()
        tech = str(row.get("tech", ""))
        wp   = str(row.get("way_policy", "")).strip().lower() if pd.notna(row.get("way_policy")) else ""
        sm   = str(row.get("static_or_migration", "")).strip().lower() if pd.notna(row.get("static_or_migration")) else ""
        rf   = str(row.get("restricted_fill", "")).strip().lower() if pd.notna(row.get("restricted_fill")) else ""
        s_val = row.get("s")
        p_val = row.get("p")
        c_val = row.get("c")
        if hca != "yes":
            if "mram" in tech and "sram" not in tech:   return "baseline_mram_only"
            elif "sram" in tech and "mram" not in tech: return "baseline_sram_only"
            elif "mram" in tech:                        return "baseline_mram_only"
            return "baseline"
        parts = []
        if wp: parts.append(wp)
        if pd.notna(s_val) and str(s_val).strip() != "":
            parts.append(f"s{int(float(s_val))}")
        parts.append("fillmram")
        if rf == "yes": parts.append("rf")
        if sm == "migration":
            if pd.notna(p_val) and str(p_val).strip() != "": parts.append(f"p{int(float(p_val))}")
            if pd.notna(c_val) and str(c_val).strip() != "": parts.append(f"c{int(float(c_val))}")
        parts.append(tech)
        return "_".join(parts)
    df["variant"] = df.apply(_synth_variant, axis=1)

    df["read_hits"]  = df["read_hits_sram"].fillna(0) + df["read_hits_mram"].fillna(0)
    df["write_hits"] = df["write_hits_sram"].fillna(0) + df["write_hits_mram"].fillna(0)
    df["misses"]     = df["dram_reads"].fillna(0) + df["dram_writes"].fillna(0)

    def _sram_ways(row):
        v = str(row["variant"])
        if "baseline_sram_only" in v: return ASSOC
        if "baseline_mram_only" in v: return 0
        s = row.get("s")
        if pd.notna(s) and str(s).strip() != "":
            try: return int(float(s))
            except: pass
        return 0
    df["sram_ways"] = df.apply(_sram_ways, axis=1).astype(int)
    df["fill_to"] = df["variant"].apply(
        lambda v: "sram" if "baseline_sram_only" in v else "mram")
    df["fill_to_l"] = df["fill_to"].str.lower()

    sm_col = df.get("static_or_migration")
    df["migration_enabled"] = (
        sm_col.fillna("").astype(str).str.strip().str.lower() == "migration"
    ) if sm_col is not None else False

    nums = ["roi_time_s", "read_hits", "write_hits", "misses", "size_mb",
            "sram_ways", "read_hits_sram", "read_hits_mram",
            "write_hits_sram", "write_hits_mram", "insn"]
    for c in nums:
        if c in df.columns: df[c] = pd.to_numeric(df[c], errors="coerce")
    df["variant"] = df["variant"].astype(str)
    df["tech"]    = df["tech"].astype(str)
    df["bench"]   = df["bench"].astype(str)

    ok = df["roi_time_s"].fillna(0) > 0
    exclude_mask = (
        df["bench"].str.contains("lbm", case=False) |
        df["bench"].str.contains("619", case=False) |
        df["bench"].str.contains("wrf", case=False) |
        df["bench"].str.contains("621", case=False)
    )
    df = df[ok & (~exclude_mask)].copy()
    print(f"After filtering: {len(df)} rows")
    return df

# =============================================================================
# Energy Calculation
# =============================================================================
def calc_energy(df):
    df = df.copy()
    df["w_sram"] = compute_w_sram(df)
    n = len(df)
    dyn_J = np.full(n, np.nan)
    leak_J = np.full(n, np.nan)

    nvsim_tech_map = {"sram14": "mram14", "sram14_mram14": "mram14"}

    for i in range(n):
        tech = df.iloc[i]["tech"]
        lookup_tech = nvsim_tech_map.get(tech, tech)
        size_mb = int(df.iloc[i]["size_mb"]) if pd.notna(df.iloc[i]["size_mb"]) else None
        if size_mb is None:
            continue

        p = NVSIM_PARAMS.get(lookup_tech, {}).get(str(size_mb))
        if p is None:
            continue

        frac = df["w_sram"].iloc[i]
        rh_s = df.iloc[i].get("read_hits_sram", 0) or 0
        rh_m = df.iloc[i].get("read_hits_mram", 0) or 0
        wh_s = df.iloc[i].get("write_hits_sram", 0) or 0
        wh_m = df.iloc[i].get("write_hits_mram", 0) or 0

        if (rh_s + rh_m + wh_s + wh_m) > 0:
            dyn_pJ = (
                rh_s * p["SRAM_R_pJ"] + rh_m * p.get("MRAM_R_pJ", 0) +
                wh_s * p["SRAM_W_pJ"] + wh_m * p.get("MRAM_W_pJ", 0)
            )
        else:
            r = df.iloc[i].get("read_hits", 0) or 0
            w = df.iloc[i].get("write_hits", 0) or 0
            dyn_pJ = (
                r * (frac * p["SRAM_R_pJ"] + (1 - frac) * p.get("MRAM_R_pJ", 0)) +
                w * (frac * p["SRAM_W_pJ"] + (1 - frac) * p.get("MRAM_W_pJ", 0))
            )

        misses = df.iloc[i].get("misses", 0) or 0
        fill_to = df.iloc[i].get("fill_to_l", "mram")
        fill_w = p["SRAM_W_pJ"] if fill_to == "sram" else p.get("MRAM_W_pJ", 0)
        dyn_pJ += misses * fill_w

        dyn_J[i] = dyn_pJ * 1e-12

        leak_mW = frac * p["SRAM_LEAK_mW"] + (1 - frac) * p.get("MRAM_LEAK_mW", 0)
        t = df.iloc[i]["roi_time_s"]
        if pd.notna(t) and t > 0:
            leak_J[i] = (leak_mW * 1e-3) * t

    df["E_dyn_J"] = dyn_J
    df["E_leak_J"] = leak_J
    df["E_total_J"] = dyn_J + leak_J
    return df

# =============================================================================
# Plot 1: Crossnode Bar
# =============================================================================
def plot_standalone_bar_crossnode(df_crossnode, sizes, save_path=None):
    names = [d[0] for d in DESIGNS]
    _DESIGN_COLORS = {
        "SRAM_7nm":  "#4878CF", "SRAM_16nm": "#e8a838", "SRAM_32nm": "#D65F5F",
        "MRAM_16nm": "#6ACC64", "MRAM_32nm": "#B07AA1",
    }
    baseline_name, baseline_size = "SRAM_7nm", 16

    b_tech, b_var = [(t, v) for n, t, v in DESIGNS if n == baseline_name][0]
    m_base = ((df_crossnode["size_mb"] == baseline_size) & (df_crossnode["tech"] == b_tech) &
              df_crossnode["variant"].str.contains(b_var, na=False))
    base_sub = df_crossnode[m_base].copy()
    base_benches = base_sub["bench"].unique()
    base_t = {b: gmean(base_sub[base_sub["bench"] == b]["roi_time_s"]) for b in base_benches}
    base_e = {b: gmean(base_sub[base_sub["bench"] == b]["E_total_J"]) for b in base_benches}

    fig, (ax_top, ax_bot) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)
    n_techs = len(DESIGNS)
    x = np.arange(len(sizes))
    w = 0.8 / n_techs

    for ti, (dname, tech, variant) in enumerate(DESIGNS):
        color = _DESIGN_COLORS.get(dname, "#999")
        speedup_means, dyn_means, leak_means = [], [], []
        for sz in sizes:
            m = ((df_crossnode["size_mb"] == sz) & (df_crossnode["tech"] == tech) &
                 df_crossnode["variant"].str.contains(variant, na=False))
            sub = df_crossnode[m]
            xs_pts, y_dyn_pts, y_leak_pts = [], [], []
            for b in base_benches:
                rows = sub[sub["bench"] == b]
                if rows.empty: continue
                dt = np.mean(rows["roi_time_s"])
                de_dyn = np.mean(rows["E_dyn_J"])
                de_leak = np.mean(rows["E_leak_J"])
                be = base_e.get(b, np.nan); bt = base_t.get(b, np.nan)
                if any(np.isnan([bt, dt, be, de_dyn, de_leak])) or bt == 0 or be == 0: continue
                xs_pts.append(bt / dt)
                y_dyn_pts.append(de_dyn / be)
                y_leak_pts.append(de_leak / be)
            if xs_pts:
                speedup_means.append(np.mean(xs_pts))
                dyn_means.append(np.mean(y_dyn_pts))
                leak_means.append(np.mean(y_leak_pts))
            else:
                speedup_means.append(0.0); dyn_means.append(0.0); leak_means.append(0.0)

        offset = (ti - n_techs / 2) * w + w / 2
        ax_top.bar(x + offset, speedup_means, w, color=color, edgecolor="black", linewidth=0.5, label=dname)
        ax_bot.bar(x + offset, dyn_means, w, color=color, alpha=0.9, edgecolor="black", linewidth=0.5)
        ax_bot.bar(x + offset, leak_means, w, bottom=dyn_means, color=color, alpha=0.3, hatch='//', edgecolor="black", linewidth=0.5)

    ax_top.axhline(1.0, color="grey", ls="--", lw=1, zorder=1)
    ax_top.set_ylabel("Speedup", fontsize=11)
    ax_top.set_title(f"Technology & Size Comparison (vs {baseline_name} {baseline_size}MB)", fontsize=12)
    ax_top.grid(axis='y', linestyle='--', alpha=0.3)
    ax_bot.axhline(1.0, color="grey", ls="--", lw=1, zorder=1)
    ax_bot.set_ylabel("Norm. L3 Energy", fontsize=11)
    ax_bot.set_xlabel("Cache Size", fontsize=11)
    ax_bot.grid(axis='y', linestyle='--', alpha=0.3)
    ax_bot.set_xticks(x)
    ax_bot.set_xticklabels([f"{sz}MB" for sz in sizes], fontsize=12)
    ax_top.legend(loc="lower center", bbox_to_anchor=(0.5, 1.10), ncol=n_techs, fontsize=10, frameon=False)
    dyn_patch = mpatches.Patch(facecolor='grey', alpha=0.9, edgecolor='black', label='Dynamic Energy')
    leak_patch = mpatches.Patch(facecolor='grey', alpha=0.3, hatch='//', edgecolor='black', label='Leakage Energy')
    ax_bot.legend(handles=[dyn_patch, leak_patch], loc="upper left", fontsize=9, framealpha=0.9)
    plt.tight_layout()
    plt.show()


# =============================================================================
# Plot 2: Static Scatter
# =============================================================================
def plot_standalone_scatter_hca(df_hybrid, sizes, save_path=None):
    hca_baseline_tech = "mram14"
    hca_baseline_var = "baseline_mram_only"
    hca_baseline_label = "All-MRAM 16nm"

    hca_benches = {}
    for sz in sizes:
        bench_sets = []
        m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == hca_baseline_tech) &
             df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())
        for _, tech, var in _STATIC_SCENARIOS:
            m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == tech) &
                 df_hybrid["variant"].str.contains(var, na=False))
            bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())
        valid = [s for s in bench_sets if s]
        hca_benches[sz] = sorted(set().union(*valid)) if valid else []

    all_hca_benches = sorted(set().union(*(hca_benches.values())))
    hca_bench_marker = {b: _BENCH_MARKERS[i % len(_BENCH_MARKERS)]
                        for i, b in enumerate(all_hca_benches)}

    nrows = len(sizes)
    fig, axes = plt.subplots(nrows, 1, figsize=(5, 3.0 * nrows))
    if nrows == 1: axes = [axes]

    for ri, sz in enumerate(sizes):
        ax = axes[ri]
        benches = hca_benches.get(sz, [])
        m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == hca_baseline_tech) &
             df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        base_sub = df_hybrid[m]
        base_t = {b: gmean(base_sub[base_sub["bench"] == b]["roi_time_s"])
                  for b in benches if not base_sub[base_sub["bench"] == b].empty}
        base_e = {b: gmean(base_sub[base_sub["bench"] == b]["E_total_J"])
                  for b in benches if not base_sub[base_sub["bench"] == b].empty}

        for si, (label, tech, var) in enumerate(_STATIC_SCENARIOS):
            m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == tech) &
                 df_hybrid["variant"].str.contains(var, na=False))
            sub = df_hybrid[m]
            color = _COMBINED_COLORS.get(label, "#999")
            xs_pts, ys_pts = [], []
            for b in benches:
                rows = sub[sub["bench"] == b]
                if rows.empty: continue
                dt = gmean(rows["roi_time_s"])
                de = gmean(rows["E_total_J"])
                bt = base_t.get(b, np.nan); be = base_e.get(b, np.nan)
                if np.isnan(bt) or np.isnan(dt) or np.isnan(be) or np.isnan(de): continue
                if bt == 0 or be == 0: continue
                ax.scatter(bt / dt, de / be, marker=hca_bench_marker[b],
                           s=60, alpha=0.95, zorder=3 + si,
                           facecolors=color, edgecolors='#444444', linewidths=1.2)
                xs_pts.append(bt / dt); ys_pts.append(de / be)
            if xs_pts:
                ax.scatter([np.mean(xs_pts)], [np.mean(ys_pts)],
                           marker="D", s=180, alpha=1.0, facecolors=color,
                           edgecolors='black', linewidths=1.5, zorder=7 + si)

        ax.axhline(1.0, color="grey", ls="--", lw=1)
        ax.axvline(1.0, color="grey", ls="--", lw=1)
        ax.set_ylim(bottom=0)
        ax.text(0.03, 0.07, f"Static {sz}MB", transform=ax.transAxes,
                ha="left", va="bottom", fontsize=11,
                bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8, edgecolor="gray"))
        ax.set_ylabel("Norm. L3 Energy", fontsize=11)
        if ri == nrows - 1:
            ax.set_xlabel(f"Speedup vs {hca_baseline_label}", fontsize=11)
        if ri == 0:
            ax.set_title("Static SRAM/MRAM Partitioning", fontsize=12)

    hca_handles = [
        Line2D([0], [0], marker="D", color="w",
               markerfacecolor=_COMBINED_COLORS.get(lab, "#999"),
               markersize=8, markeredgecolor="k", markeredgewidth=1.0, label=lab)
        for lab in ["All-MRAM 16nm", "All-SRAM 16nm", "Static 4 SRAM", "Static 8 SRAM", "Static 12 SRAM"]
    ]
    bench_handles = [
        Line2D([0], [0], marker=hca_bench_marker[b], color="w",
               markerfacecolor="grey", markersize=6, markeredgecolor="k", markeredgewidth=0.5,
               label=_short_bench(b))
        for b in all_hca_benches
    ]

    plt.tight_layout()
    plt.subplots_adjust(top=0.90 if nrows > 1 else 0.82)
    fig.legend(handles=hca_handles + bench_handles, loc="upper center",
               bbox_to_anchor=(0.5, 1.02), ncol=5, frameon=False, fontsize=8)
    plt.show()



def plot_read_only_sweep(df, sizes=[16, 32, 128], save_path=None):
    """Bar chart: one row per benchmark.
    X = cache size, bars colored by MRAM read multiplier.
    Y = Static S4 speedup vs All-MRAM.

    Matches hca.py behavior:
      - baseline matched by both size and rd_mult
      - sweep subset includes all rd_mult rows, then filters Static S4 in-loop
      - uses first row values rather than gmean
      - optionally saves figure if save_path is given
    """
    _SHOW_BENCHES = ["mcf", "omnetpp"]

    _VARIANT_MAP = [
        ("baseline_mram_only",            "All-MRAM"),
        ("noparity_s4_fillmram_rf_p1_c0", None),
        ("noparity_s4_fillmram_rf",       None),
        ("noparity_s4_fillmram_p1_c0",    None),
        ("noparity_s4_fillmram",          "Static S4"),
    ]

    _MULT_COLORS = {2: "#1f77b4", 3: "#ff7f0e", 4: "#2ca02c", 5: "#d62728"}

    def _classify(var_str):
        for substr, label in _VARIANT_MAP:
            if substr in str(var_str):
                return label
        return None

    def _read_mult(tag):
        m = re.match(r"r?(\d+)x", str(tag))
        return int(m.group(1)) if m else None

    df = df.copy()
    df["design"] = df["variant"].apply(_classify)
    df = df[df["design"].notna()].copy()

    df["rd_mult"] = df["rw_tag"].apply(_read_mult)
    df["bench_short"] = df["bench"].apply(_short_bench)
    df["sz_mb"] = df["size_mb"]
    df["elapsed_fs"] = df["elapsed_ns"] * 1e6  # ns -> fs

    df = df[df["bench_short"].isin(_SHOW_BENCHES)].copy()

    # Separate baseline from sweep exactly like hca.py
    df_base = df[df["design"] == "All-MRAM"].copy()
    df_sweep = df[df["rd_mult"].notna()].copy()

    rd_mults = sorted(df_sweep["rd_mult"].dropna().unique())

    nrows = len(_SHOW_BENCHES)
    fig, axes = plt.subplots(nrows, 1, figsize=(5, 2.4 * nrows), squeeze=False)

    for ri, bench in enumerate(_SHOW_BENCHES):
        ax = axes[ri, 0]
        b_base = df_base[df_base["bench_short"] == bench]
        b_sweep = df_sweep[df_sweep["bench_short"] == bench]

        x = np.arange(len(sizes))
        n_m = len(rd_mults)
        w = 0.7 / n_m if n_m > 0 else 0.7

        for mi, mult in enumerate(rd_mults):
            vals = []
            for sz in sizes:
                # Match hca.py exactly: baseline must match BOTH size and multiplier
                base_sub = b_base[
                    (b_base["sz_mb"] == sz) &
                    (b_base["rd_mult"] == mult)
                ]
                s4_sub = b_sweep[
                    (b_sweep["design"] == "Static S4") &
                    (b_sweep["rd_mult"] == mult) &
                    (b_sweep["sz_mb"] == sz)
                ]

                if base_sub.empty or s4_sub.empty:
                    vals.append(np.nan)
                else:
                    bt = base_sub["elapsed_fs"].values[0]
                    st = s4_sub["elapsed_fs"].values[0]
                    su = bt / st if st > 0 else np.nan
                    vals.append(su)
                    print(f"    {bench} {sz}MB {mult}x: speedup={su:.4f}")

            offset = (mi - n_m / 2) * w + w / 2
            ax.bar(
                x + offset, vals, w,
                color=_MULT_COLORS.get(mult, "#999"),
                edgecolor="white",
                label=f"{mult}x" if ri == 0 else None
            )

        ax.axhline(1.0, color="grey", ls="--", lw=1, zorder=0)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{sz}MB" for sz in sizes], fontsize=10)
        ax.set_ylabel("Speedup vs All-MRAM", fontsize=10)
        ax.set_title(bench, fontsize=11)
        ax.grid(axis="y", linestyle="--", alpha=0.3)
        ax.set_ylim(bottom=0.0)

    h, l = axes[0, 0].get_legend_handles_labels()
    plt.tight_layout()
    plt.subplots_adjust(top=0.84, hspace=0.40)
    fig.legend(
        h, l,
        loc="upper center", bbox_to_anchor=(0.5, 0.98),
        ncol=max(len(rd_mults), 1),
        frameon=False, fontsize=9,
        title="MRAM Read Multiplier", title_fontsize=9
    )

    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")

    plt.show()
    
    
if __name__ == "__main__":
    df_all = load_all()
    print(f"\nAll techs:  {sorted(df_all['tech'].unique())}")
    print(f"All sizes:  {sorted(df_all['size_mb'].dropna().unique())}")

    # Crossnode slice (hca=no, no lat_scale)
    is_baseline = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "no") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_cross = calc_energy(df_all[is_baseline].copy())
    print(f"\nCrossnode: {len(df_cross)} rows")

    sizes = [sz for sz in sorted(df_cross["size_mb"].dropna().unique().astype(int))
             if sz in (16, 32, 128)]
    print(f"Sizes: {sizes}")

    # HCA slice (hca=yes, no lat_scale)
    is_hca = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "yes") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_hca = calc_energy(df_all[is_hca].copy())
    print(f"HCA: {len(df_hca)} rows")

    # Hybrid = HCA rows + MRAM14/SRAM14 baselines
    df_baselines = df_cross[
        df_cross["tech"].isin(["mram14", "sram14"]) &
        (
            df_cross["variant"].str.contains("baseline_mram_only", na=False) |
            df_cross["variant"].str.contains("baseline_sram_only", na=False)
        )
    ].copy()
    df_hybrid = pd.concat([df_hca, df_baselines], ignore_index=True)
    print(f"Hybrid: {len(df_hybrid)} rows")

    # Sweep slice (rows with read-latency tag + matching All-MRAM baselines)
    df_sweep_hca = df_all[df_all["rw_tag"].fillna("") != ""].copy()
    if not df_sweep_hca.empty:
        sweep_benches = df_sweep_hca["bench"].unique()
        sweep_sizes = df_sweep_hca["size_mb"].unique()

        df_sweep_base = df_all[
            df_all["variant"].str.contains("baseline_mram_only", na=False) &
            df_all["bench"].isin(sweep_benches) &
            df_all["size_mb"].isin(sweep_sizes)
        ].copy()

        df_sweep = pd.concat([df_sweep_hca, df_sweep_base], ignore_index=True)
    else:
        df_sweep = df_sweep_hca

    print("\n=== Crossnode Bar ===")
    plot_standalone_bar_crossnode(df_cross, sizes=sizes)

    print("\n=== Static Scatter ===")
    plot_standalone_scatter_hca(df_hybrid, sizes=sizes)

    if not df_sweep.empty:
        print(f"\n=== Read-Only Latency Sweep ({len(df_sweep)} rows) ===")
        plot_read_only_sweep(df_sweep, sizes=[16, 32, 128])
    else:
        print("\n[SKIP] No latency sweep rows found")

    print("\nDone!")
```

```python
import csv
from pathlib import Path

CSV_PATH = Path("~/COSC_498/miniMXE/repro/agg/hca_workload.csv").expanduser()

WKLDS_REP = ["mcf", "perlbench", "omnetpp", "deepsjeng"]

def load_hca_workload_csv(csv_path):
    data = {}
    with open(csv_path, newline="") as f:
        for row in csv.DictReader(f):
            wl = row["workload"]
            sz = int(row["size_mb"])
            cfg = row["cfg"]
            data[(wl, sz, cfg)] = {
                "elapsed_ns": float(row["elapsed_ns"]) if row["elapsed_ns"] else None,
                "l3_load_misses": float(row["l3_load_misses"]) if row["l3_load_misses"] else None,
                "mram_write_bytes": float(row["mram_write_bytes"]) if row["mram_write_bytes"] else None,
                "hybrid_promotions": float(row["hybrid_promotions"]) if row["hybrid_promotions"] else 0.0,
            }
    return data

def print_table6_like_image(hca_data):
    configs = [("MRAM", "MRAM"), ("S4", "S4"), ("P4C32", "P4C32"), ("P1C0", "P1C0")]

    print("=" * 86)
    print("TABLE 6")
    print("Representative 16 MB workloads under unrestricted fill")
    print("=" * 86)
    print(f"{'Workload':<12} {'Cfg':<8} {'Perf':>8} {'L3 miss':>10} {'MRAM wr':>10} {'Promo':>10}")
    print("-" * 86)

    for wl in WKLDS_REP:
        base = hca_data.get((wl, 16, "MRAM"))
        if not base:
            continue

        b_ns = base["elapsed_ns"]
        b_miss = base["l3_load_misses"]
        b_mwb = base["mram_write_bytes"]

        for label, cfg in configs:
            r = hca_data.get((wl, 16, cfg))
            if not r:
                continue

            perf = b_ns / r["elapsed_ns"] if r["elapsed_ns"] else None
            l3_miss = r["l3_load_misses"] / b_miss if b_miss else None
            mram_wr = r["mram_write_bytes"] / b_mwb if b_mwb else None
            promo = int(r["hybrid_promotions"])

            wl_disp = wl if label == "MRAM" else ""
            print(f"{wl_disp:<12} {label:<8} {perf:>8.3f} {l3_miss:>10.3f} {mram_wr:>10.3f} {promo:>10,d}")

        print("-" * 86)

def print_table7_like_image(hca_data):
    print("=" * 110)
    print("TABLE 7")
    print("Restricted-fill migration summary")
    print("=" * 110)

    hdr1 = f"{'Workload':<12} | {'16 MB':^30} | {'32 MB':^30} | {'128 MB':^22}"
    hdr2 = (
        f"{'':<12} | {'S4-RF':>7} {'P4C32/S':>8} {'P1C0/S':>8} {'P1C0 mr':>7} |"
        f" {'S4-RF':>7} {'P4C32/S':>8} {'P1C0/S':>8} {'P1C0 mr':>7} |"
        f" {'S4-RF':>7} {'P4C32/S':>8} {'P1C0/S':>8}"
    )
    print(hdr1)
    print(hdr2)
    print("-" * 110)

    for wl in WKLDS_REP:
        line = f"{wl:<12} |"
        for sz in [16, 32, 128]:
            base = hca_data.get((wl, sz, "MRAM"))
            st   = hca_data.get((wl, sz, "S4_RF"))
            m4   = hca_data.get((wl, sz, "P4C32_RF"))
            m1   = hca_data.get((wl, sz, "P1C0_RF"))

            if not all([base, st, m4, m1]):
                if sz != 128:
                    line += f" {'—':>7} {'—':>8} {'—':>8} {'—':>7} |"
                else:
                    line += f" {'—':>7} {'—':>8} {'—':>8}"
                continue

            s4rf = base["elapsed_ns"] / st["elapsed_ns"] if st["elapsed_ns"] else None
            p4c  = st["elapsed_ns"] / m4["elapsed_ns"] if m4["elapsed_ns"] else None
            p1c  = st["elapsed_ns"] / m1["elapsed_ns"] if m1["elapsed_ns"] else None

            if sz != 128:
                mr = (
                    m1["l3_load_misses"] / base["l3_load_misses"]
                    if base["l3_load_misses"] else None
                )
                line += f" {s4rf:>7.3f} {p4c:>8.3f} {p1c:>8.3f} {mr:>6.2f}x |"
            else:
                line += f" {s4rf:>7.3f} {p4c:>8.3f} {p1c:>8.3f}"

        print(line)

hca_data = load_hca_workload_csv(CSV_PATH)
print_table6_like_image(hca_data)
print()
print_table7_like_image(hca_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

import csv
import os

CSV_PATH = os.path.expanduser("~/COSC_498/miniMXE/repro/agg/selective_dvfs.csv")

label_map = {
    ("500.perlbench_r", "n1"): "perl",
    ("502.gcc_r", "n1"): "gcc",
    ("505.mcf_r", "n1"): "mcf",
    ("520.omnetpp_r", "n1"): "omnetpp",
    ("523.xalancbmk_r", "n1"): "xalan",
    ("531.deepsjeng_r", "n1"): "dsj",
    ("541.leela_r", "n1"): "leela",
    ("557.xz_r", "n1"): "xz",
    ("648.exchange2_s", "n1"): "exch",
    ("649.fotonik3d_s", "n1"): "foto",

    ("502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n4"): "gcc×4",
    ("505.mcf_r+500.perlbench_r+648.exchange2_s+649.fotonik3d_s", "n4"): "mcf+perl+exch+foto",
    ("505.mcf_r+505.mcf_r+502.gcc_r+502.gcc_r", "n4"): "mcf×2+gcc×2",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r", "n4"): "mcf×4",
    ("523.xalancbmk_r+523.xalancbmk_r+502.gcc_r+502.gcc_r", "n4"): "xalan×2+gcc×2",

    ("502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "gcc×8",
    ("505.mcf_r+505.mcf_r+500.perlbench_r+500.perlbench_r+648.exchange2_s+648.exchange2_s+649.fotonik3d_s+649.fotonik3d_s", "n8"): "mcf×2+perl×2+exch×2+foto×2",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "mcf×4+gcc×4",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r", "n8"): "mcf×8",
    ("523.xalancbmk_r+523.xalancbmk_r+523.xalancbmk_r+523.xalancbmk_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "xalan×4+gcc×4",
}

data = []

with open(CSV_PATH, newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row["size_mb"] != "16":
            continue

        key = (row["workload"], row["n_tag"])
        if key not in label_map:
            continue

        label = label_map[key]

        if row["n_tag"] == "n1":
            dvfs = float(row["speedup_vs_sram7_ratio"])
            mram = dvfs / float(row["speedup_vs_mram14_ratio"])
        else:
            dvfs = float(row["wsn_vs_sram7"])
            mram = float(row["wsn_vs_mram14"]) / dvfs

        data.append((label, round(dvfs, 4), 1.0000, round(mram, 4)))

df = pd.DataFrame(data, columns=["Workload", "DVFS", "SRAM7", "MRAM14"])

def total_n(workload: str) -> int:
    total = 0
    for token in workload.split("+"):
        m = re.search(r"×(\d+)", token)
        total += int(m.group(1)) if m else 1
    return total

def shorten_label(label: str) -> str:
    return label.replace("omnetpp", "omnet")

def smart_multiline_label(label: str, max_parts_per_line=2):
    parts = shorten_label(label).split("+")
    lines = []
    for i in range(0, len(parts), max_parts_per_line):
        chunk = parts[i:i + max_parts_per_line]
        line = "+".join(chunk)
        if i > 0:
            line = "+" + line
        lines.append(line)
    return "\n".join(lines)

df["N"] = df["Workload"].apply(total_n)
df["MRAM14_rel"] = df["MRAM14"] / df["SRAM7"]
df["DVFS_rel"] = df["DVFS"] / df["SRAM7"]
df["Label"] = df["Workload"].apply(smart_multiline_label)

# sort within each regime by DVFS, then concatenate regimes in order
groups = []
boundaries = []
cursor = 0

for n in [1, 4, 8]:
    g = (
        df[df["N"] == n]
        .sort_values("DVFS_rel", ascending=False)
        .reset_index(drop=True)
    )
    groups.append(g)
    cursor += len(g)
    boundaries.append(cursor - 0.5)

plot_df = pd.concat(groups, ignore_index=True)
summary_df = plot_df[["Label", "N", "DVFS_rel", "MRAM14_rel"]].copy()
summary_df = summary_df.rename(columns={
    "Label": "Workload",
    "DVFS_rel": "DVFS vs SRAM7",
    "MRAM14_rel": "MRAM14 vs SRAM7",
})

summary_df["DVFS vs SRAM7"] = summary_df["DVFS vs SRAM7"].map(lambda x: f"{x:.4f}")
summary_df["MRAM14 vs SRAM7"] = summary_df["MRAM14 vs SRAM7"].map(lambda x: f"{x:.4f}")

print("\nSummary table for plotted points:\n")
print(summary_df.to_string(index=False))

x = list(range(len(plot_df)))

y_min = min(plot_df["MRAM14_rel"].min(), plot_df["DVFS_rel"].min())
y_max = max(plot_df["MRAM14_rel"].max(), plot_df["DVFS_rel"].max())
ylim = (min(0.95, y_min - 0.01), y_max + 0.015)

fig, ax = plt.subplots(figsize=(14, 4.5))

ax.plot(x, plot_df["MRAM14_rel"], "o", label="MRAM16")
ax.plot(x, plot_df["DVFS_rel"], "s", label="MRAM16 + DVFS")
ax.axhline(1.0, linestyle="--", linewidth=1)

# regime separators between N=1|4 and N=4|8
for boundary in boundaries[:-1]:
    ax.axvline(boundary, linestyle=":", linewidth=1)

ax.set_xticks(x)
ax.set_xticklabels(plot_df["Label"], rotation=45, ha="right")
ax.set_ylim(*ylim)
ax.set_ylabel("Speedup vs SRAM7")
ax.set_xlabel("Workload / mix")
ax.grid(axis="y", alpha=0.25)
ax.legend(frameon=False, ncol=2, loc="upper right")

# optional regime labels above groups
group_sizes = [len(g) for g in groups]
starts = [0, group_sizes[0], group_sizes[0] + group_sizes[1]]
centers = [
    starts[0] + (group_sizes[0] - 1) / 2,
    starts[1] + (group_sizes[1] - 1) / 2,
    starts[2] + (group_sizes[2] - 1) / 2,
]
for center, label in zip(centers, ["N=1", "N=4", "N=8"]):
    ax.text(center, 1.01, label, transform=ax.get_xaxis_transform(),
            ha="center", va="bottom")

plt.tight_layout()
plt.savefig("workload_one_figure_grouped.png", dpi=240)
plt.show()
```

```python
import csv
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

CSV_PATH = Path("/home/skataoka26/COSC_498/miniMXE/repro/agg/selective_dvfs.csv")
SIZE_MB = 16  # change if needed

def parse_vec(s):
    if not s or (isinstance(s, float) and np.isnan(s)):
        return []
    return [float(x) for x in str(s).split("|") if str(x).strip()]

# -----------------------------
# Load CSV
# -----------------------------
df = pd.read_csv(CSV_PATH)

# Only multicore mixes at the target LLC size
df = df[(df["size_mb"] == SIZE_MB) & (df["n_tag"].isin(["n4", "n8"]))].copy()

if df.empty:
    raise RuntimeError(f"No rows found for size_mb={SIZE_MB}")

# -----------------------------
# Build positive-gain shares
# g_i = max(0, ws_term_i - 1)
# p_i = g_i / sum_j g_j
# -----------------------------
records = []

for _, row in df.iterrows():
    terms = parse_vec(row["ws_terms_vs_sram7"])
    if not terms:
        continue

    gains = np.array([max(0.0, x - 1.0) for x in terms], dtype=float)
    total_gain = gains.sum()

    # If there is no positive gain at all, skip that mix
    if total_gain <= 0:
        continue

    shares = np.sort(gains / total_gain)[::-1]   # descending
    cumshares = np.cumsum(shares)

    records.append({
        "n_tag": row["n_tag"],
        "workload_short": row["workload_short"],
        "n_cores": int(row["n_cores"]),
        "shares": shares,
        "cumshares": cumshares,
        "wsn_vs_sram7": float(row["wsn_vs_sram7"]),
    })

if not records:
    raise RuntimeError("No valid multicore rows with positive gain found.")

# -----------------------------
# Option A: concentration curves
# average cumulative concentration by group
# -----------------------------
grouped = {}
for tag in ["n4", "n8"]:
    sub = [r for r in records if r["n_tag"] == tag]
    if not sub:
        continue

    max_len = max(len(r["cumshares"]) for r in sub)

    # Pad with 1.0 after the cumulative curve has finished
    arr = np.ones((len(sub), max_len), dtype=float)
    for i, r in enumerate(sub):
        arr[i, :len(r["cumshares"])] = r["cumshares"]

    grouped[tag] = {
        "mean_cum": arr.mean(axis=0),
        "std_cum": arr.std(axis=0),
        "n_mixes": len(sub),
    }

plt.figure(figsize=(7, 3.))

label_map = {"n4": "4-core", "n8": "8-core"}

for tag in ["n4", "n8"]:
    if tag not in grouped:
        continue
    y = grouped[tag]["mean_cum"]
    x = np.arange(1, len(y) + 1)
    plt.plot(x, y, marker="o", label=label_map[tag])
    plt.fill_between(
        x,
        np.clip(y - grouped[tag]["std_cum"], 0, 1),
        np.clip(y + grouped[tag]["std_cum"], 0, 1),
        alpha=0.15,
    )

plt.xlabel("Top k most-benefited co-runners")
plt.ylabel("Cum. share of total positive gain")
plt.xticks(range(1, max(len(v["mean_cum"]) for v in grouped.values()) + 1))
plt.ylim(0, 1.02)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# -----------------------------
# Option B: top-1 / top-2 share table
# -----------------------------
rows = []
for tag in ["n4", "n8"]:
    sub = [r for r in records if r["n_tag"] == tag]
    if not sub:
        continue

    top1 = []
    top2 = []
    top3 = []
    mean_perf = []

    for r in sub:
        shares = r["shares"]
        top1.append(shares[0] if len(shares) >= 1 else np.nan)
        top2.append(shares[:2].sum() if len(shares) >= 2 else shares.sum())
        top3.append(shares[:3].sum() if len(shares) >= 3 else shares.sum())
        mean_perf.append(r["wsn_vs_sram7"])

    rows.append({
        "Case": label_map[tag],
        "Mixes": len(sub),
        "MeanPerf": np.mean(mean_perf),
        "Top1Share": np.mean(top1),
        "Top2Share": np.mean(top2),
        "Top3Share": np.mean(top3),
    })

summary = pd.DataFrame(rows)
summary["MeanPerf"] = summary["MeanPerf"].map(lambda x: f"{x:.4f}x")
summary["Top1Share"] = summary["Top1Share"].map(lambda x: f"{100*x:.1f}%")
summary["Top2Share"] = summary["Top2Share"].map(lambda x: f"{100*x:.1f}%")
summary["Top3Share"] = summary["Top3Share"].map(lambda x: f"{100*x:.1f}%")

display(summary)

# Optional: show per-mix details underneath
detail_rows = []
for r in records:
    shares = r["shares"]
    detail_rows.append({
        "Case": label_map[r["n_tag"]],
        "Workload": r["workload_short"],
        "Perf": f"{r['wsn_vs_sram7']:.4f}x",
        "Top1Share": f"{100*(shares[0] if len(shares) >= 1 else 0.0):.1f}%",
        "Top2Share": f"{100*(shares[:2].sum() if len(shares) >= 2 else shares.sum()):.1f}%",
        "Top3Share": f"{100*(shares[:3].sum() if len(shares) >= 3 else shares.sum()):.1f}%",
    })

detail_df = pd.DataFrame(detail_rows).sort_values(["Case", "Workload"])
display(detail_df)
```

```python
import os
import sqlite3
import csv
from io import StringIO

BASE_FREQ_GHZ = 2.2
RUNS_DIR = "/home/skataoka26/COSC_498/miniMXE/repro/hca/hca_sunnycove/2_cross_node/mram14/runs"

def get_stat(cur, obj, metric, prefix, core):
    cur.execute("""
        SELECT v.value
        FROM "values" v
        JOIN names n ON v.nameid = n.nameid
        JOIN prefixes p ON v.prefixid = p.prefixid
        WHERE n.objectname=? AND n.metricname=? AND p.prefixname=? AND v.core=?
    """, (obj, metric, prefix, core))
    row = cur.fetchone()
    return float(row[0]) if row else None

def get_delta(cur, obj, metric, core):
    b = get_stat(cur, obj, metric, "roi-begin", core)
    e = get_stat(cur, obj, metric, "roi-end", core)
    if b is not None and e is not None:
        return e - b
    return e

def safe_div(a, b):
    if a is None or b is None or b == 0:
        return None
    return a / b

def fmt(x, nd=6):
    if x is None:
        return "NA"
    return f"{x:.{nd}f}"

dbs = sorted([
    os.path.join(RUNS_DIR, d, "sz16M", "baseline_mram_only_mram14", "sim.stats.sqlite3")
    for d in os.listdir(RUNS_DIR)
    if os.path.isdir(os.path.join(RUNS_DIR, d))
    and os.path.exists(os.path.join(RUNS_DIR, d, "sz16M", "baseline_mram_only_mram14", "sim.stats.sqlite3"))
])

buf = StringIO()
w = csv.writer(buf)
w.writerow([
    "workload",
    "instr",
    "llc_access_pkI",
    "llc_miss_pkI",
    "base_frac",
    "branch_frac",
    "l3_frac",
    "dram_frac",
    "ipc",
    "mlp_proxy_1",
    "mlp_proxy_2",
])

for db in dbs:
    workload = os.path.basename(os.path.dirname(os.path.dirname(os.path.dirname(db))))
    try:
        try:
            con = sqlite3.connect(f"file:{db}?mode=ro", uri=True)
        except Exception:
            con = sqlite3.connect(db)
        cur = con.cursor()

        core = 0
        instr = get_delta(cur, "performance_model", "instruction_count", core)
        if instr is None or instr <= 0:
            instr = get_delta(cur, "core", "instructions", core)

        nonidle = get_delta(cur, "thread", "nonidle_elapsed_time", core)

        base = get_delta(cur, "rob_timer", "cpiBase", core)
        branch = get_delta(cur, "rob_timer", "cpiBranchPredictor", core)
        l3 = get_delta(cur, "rob_timer", "cpiDataCacheL3", core)
        dram = get_delta(cur, "rob_timer", "cpiDataCachedram", core)
        if dram is None:
            dram = 0.0

        ollc = get_delta(cur, "rob_timer", "outstandingLongLatencyCycles", core)
        olli = get_delta(cur, "rob_timer", "outstandingLongLatencyInsns", core)

        l3_misses = get_delta(cur, "L3", "l3_misses", core)
        if l3_misses is None:
            lm = get_delta(cur, "L3", "load-misses", core) or 0.0
            sm = get_delta(cur, "L3", "store-misses", core) or 0.0
            l3_misses = lm + sm

        l3_read_hits = get_delta(cur, "L3", "l3_read_hits", core)
        l3_write_hits = get_delta(cur, "L3", "l3_write_hits", core)

        llc_accesses = None
        if l3_read_hits is not None and l3_write_hits is not None and l3_misses is not None:
            llc_accesses = l3_read_hits + l3_write_hits + l3_misses
        else:
            loads = get_delta(cur, "L3", "loads", core)
            stores = get_delta(cur, "L3", "stores", core)
            if loads is not None and stores is not None:
                llc_accesses = loads + stores

        llc_access_pki = safe_div(llc_accesses * 1000.0, instr) if llc_accesses is not None else None
        llc_miss_pki = safe_div(l3_misses * 1000.0, instr) if l3_misses is not None else None

        base_frac = safe_div(base, nonidle)
        branch_frac = safe_div(branch, nonidle)
        l3_frac = safe_div(l3, nonidle)
        dram_frac = safe_div(dram, nonidle)

        ipc = None
        if instr is not None and nonidle is not None and nonidle > 0:
            approx_cycles = nonidle * BASE_FREQ_GHZ * 1e9 / 1e15
            ipc = safe_div(instr, approx_cycles)

        mlp_proxy_1 = safe_div(ollc, nonidle)
        mlp_proxy_2 = safe_div(ollc, olli)

        w.writerow([
            workload,
            fmt(instr, 0),
            fmt(llc_access_pki, 3),
            fmt(llc_miss_pki, 3),
            fmt(base_frac, 6),
            fmt(branch_frac, 6),
            fmt(l3_frac, 6),
            fmt(dram_frac, 6),
            fmt(ipc, 3),
            fmt(mlp_proxy_1, 6),
            fmt(mlp_proxy_2, 6),
        ])

        con.close()
    except Exception as e:
        w.writerow([workload, "ERROR", str(e)])

print(buf.getvalue())
```

```python
def plot_standalone_bar_crossnode(df_crossnode, sizes, save_path=None):
    _DESIGN_COLORS = {
        "SRAM_7nm":  "#4878CF",
        "SRAM_16nm": "#e8a838",
        "SRAM_32nm": "#D65F5F",
        "MRAM_16nm": "#6ACC64",
        "MRAM_32nm": "#B07AA1",
    }
    baseline_name, baseline_size = "SRAM_7nm", 16

    b_tech, b_var = [(t, v) for n, t, v in DESIGNS if n == baseline_name][0]
    m_base = (
        (df_crossnode["size_mb"] == baseline_size) &
        (df_crossnode["tech"] == b_tech) &
        (df_crossnode["variant"].str.contains(b_var, na=False))
    )
    base_sub = df_crossnode[m_base].copy()
    base_benches = base_sub["bench"].unique()
    base_t = {b: gmean(base_sub[base_sub["bench"] == b]["roi_time_s"]) for b in base_benches}
    base_e = {b: gmean(base_sub[base_sub["bench"] == b]["E_total_J"]) for b in base_benches}

    fig, (ax_top, ax_bot) = plt.subplots(1, 2, figsize=(12, 4.8))

    n_techs = len(DESIGNS)
    x = np.arange(len(sizes))
    w = 0.8 / n_techs

    for ti, (dname, tech, variant) in enumerate(DESIGNS):
        color = _DESIGN_COLORS.get(dname, "#999999")
        speedup_means, dyn_means, leak_means = [], [], []

        for sz in sizes:
            m = (
                (df_crossnode["size_mb"] == sz) &
                (df_crossnode["tech"] == tech) &
                (df_crossnode["variant"].str.contains(variant, na=False))
            )
            sub = df_crossnode[m]

            xs_pts, y_dyn_pts, y_leak_pts = [], [], []
            for b in base_benches:
                rows = sub[sub["bench"] == b]
                if rows.empty:
                    continue

                dt = np.mean(rows["roi_time_s"])
                de_dyn = np.mean(rows["E_dyn_J"])
                de_leak = np.mean(rows["E_leak_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)

                if any(np.isnan([bt, dt, be, de_dyn, de_leak])) or bt == 0 or be == 0:
                    continue

                xs_pts.append(bt / dt)
                y_dyn_pts.append(de_dyn / be)
                y_leak_pts.append(de_leak / be)

            if xs_pts:
                speedup_means.append(np.mean(xs_pts))
                dyn_means.append(np.mean(y_dyn_pts))
                leak_means.append(np.mean(y_leak_pts))
            else:
                speedup_means.append(0.0)
                dyn_means.append(0.0)
                leak_means.append(0.0)

        offset = (ti - n_techs / 2) * w + w / 2
        ax_top.bar(
            x + offset, speedup_means, w,
            color=color, edgecolor="black", linewidth=0.5, label=dname
        )
        ax_bot.bar(
            x + offset, dyn_means, w,
            color=color, alpha=0.9, edgecolor="black", linewidth=0.5
        )
        ax_bot.bar(
            x + offset, leak_means, w, bottom=dyn_means,
            color=color, alpha=0.3, hatch='//',
            edgecolor="black", linewidth=0.5
        )

    ax_top.axhline(1.0, color="grey", ls="--", lw=1, zorder=1)
    ax_top.set_ylabel("Speedup", fontsize=11)
    ax_top.set_xlabel("Cache Size", fontsize=11)
    ax_top.grid(axis='y', linestyle='--', alpha=0.3)
    ax_top.set_xticks(x)
    ax_top.set_xticklabels([f"{sz}MB" for sz in sizes], fontsize=12)

    ax_bot.axhline(1.0, color="grey", ls="--", lw=1, zorder=1)
    ax_bot.set_ylabel("Norm. L3 Energy", fontsize=11)
    ax_bot.set_xlabel("Cache Size", fontsize=11)
    ax_bot.grid(axis='y', linestyle='--', alpha=0.3)
    ax_bot.set_xticks(x)
    ax_bot.set_xticklabels([f"{sz}MB" for sz in sizes], fontsize=12)

    design_handles, design_labels = ax_top.get_legend_handles_labels()
    fig.legend(
        design_handles, design_labels,
        loc="upper center",
        bbox_to_anchor=(0.5, 0.98),
        ncol=len(DESIGNS),
        frameon=False,
        fontsize=10
    )

    dyn_patch = mpatches.Patch(
        facecolor='grey', alpha=0.9, edgecolor='black', label='Dynamic Energy'
    )
    leak_patch = mpatches.Patch(
        facecolor='grey', alpha=0.3, hatch='//', edgecolor='black', label='Leakage Energy'
    )
    ax_bot.legend(
        handles=[dyn_patch, leak_patch],
        loc="upper left",
        fontsize=9,
        framealpha=0.9
    )

    plt.tight_layout()
    plt.subplots_adjust(top=0.84)
    plt.show()


# =============================================================================
# Plot 2: Static Scatter
# =============================================================================
def plot_standalone_scatter_hca(df_hybrid, sizes, save_path=None):
    hca_baseline_tech = "mram14"
    hca_baseline_var = "baseline_mram_only"
    hca_baseline_label = "All-MRAM 16nm"

    hca_benches = {}
    for sz in sizes:
        bench_sets = []
        m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == hca_baseline_tech) &
             df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())
        for _, tech, var in _STATIC_SCENARIOS:
            m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == tech) &
                 df_hybrid["variant"].str.contains(var, na=False))
            bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())
        valid = [s for s in bench_sets if s]
        hca_benches[sz] = sorted(set().union(*valid)) if valid else []

    all_hca_benches = sorted(set().union(*(hca_benches.values())))
    hca_bench_marker = {b: _BENCH_MARKERS[i % len(_BENCH_MARKERS)]
                        for i, b in enumerate(all_hca_benches)}

    # horizontal: 1 row, N cols
    ncols = len(sizes)
    fig, axes = plt.subplots(1, ncols, figsize=(4.8 * ncols, 4.2))
    if ncols == 1:
        axes = [axes]

    for ci, sz in enumerate(sizes):
        ax = axes[ci]
        benches = hca_benches.get(sz, [])
        m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == hca_baseline_tech) &
             df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        base_sub = df_hybrid[m]
        base_t = {b: gmean(base_sub[base_sub["bench"] == b]["roi_time_s"])
                  for b in benches if not base_sub[base_sub["bench"] == b].empty}
        base_e = {b: gmean(base_sub[base_sub["bench"] == b]["E_total_J"])
                  for b in benches if not base_sub[base_sub["bench"] == b].empty}

        for si, (label, tech, var) in enumerate(_STATIC_SCENARIOS):
            m = ((df_hybrid["size_mb"] == sz) & (df_hybrid["tech"] == tech) &
                 df_hybrid["variant"].str.contains(var, na=False))
            sub = df_hybrid[m]
            color = _COMBINED_COLORS.get(label, "#999")
            xs_pts, ys_pts = [], []
            for b in benches:
                rows = sub[sub["bench"] == b]
                if rows.empty:
                    continue
                dt = gmean(rows["roi_time_s"])
                de = gmean(rows["E_total_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)
                if np.isnan(bt) or np.isnan(dt) or np.isnan(be) or np.isnan(de):
                    continue
                if bt == 0 or be == 0:
                    continue
                ax.scatter(bt / dt, de / be, marker=hca_bench_marker[b],
                           s=60, alpha=0.95, zorder=3 + si,
                           facecolors=color, edgecolors='#444444', linewidths=1.2)
                xs_pts.append(bt / dt)
                ys_pts.append(de / be)
            if xs_pts:
                ax.scatter([np.mean(xs_pts)], [np.mean(ys_pts)],
                           marker="D", s=180, alpha=1.0, facecolors=color,
                           edgecolors='black', linewidths=1.5, zorder=7 + si)

        ax.axhline(1.0, color="grey", ls="--", lw=1)
        ax.axvline(1.0, color="grey", ls="--", lw=1)
        ax.set_ylim(bottom=0)
        ax.set_title(f"Static {sz}MB", fontsize=11)
        ax.set_xlabel(f"Speedup vs {hca_baseline_label}", fontsize=10)
        if ci == 0:
            ax.set_ylabel("Norm. L3 Energy", fontsize=11)

    hca_handles = [
        Line2D([0], [0], marker="D", color="w",
               markerfacecolor=_COMBINED_COLORS.get(lab, "#999"),
               markersize=8, markeredgecolor="k", markeredgewidth=1.0, label=lab)
        for lab in ["All-MRAM 16nm", "All-SRAM 16nm", "Static 4 SRAM", "Static 8 SRAM", "Static 12 SRAM"]
    ]
    bench_handles = [
        Line2D([0], [0], marker=hca_bench_marker[b], color="w",
               markerfacecolor="grey", markersize=6, markeredgecolor="k", markeredgewidth=0.5,
               label=_short_bench(b))
        for b in all_hca_benches
    ]

    plt.tight_layout()
    plt.subplots_adjust(top=0.80)
    fig.legend(handles=hca_handles + bench_handles, loc="upper center",
               bbox_to_anchor=(0.5, 1.02), ncol=5, frameon=False, fontsize=8)
    plt.show()


def plot_read_only_sweep(df, sizes=[16, 32, 128], save_path=None):
    """Bar chart: one panel per benchmark, laid out horizontally."""
    _SHOW_BENCHES = ["mcf", "omnetpp"]

    _VARIANT_MAP = [
        ("baseline_mram_only",            "All-MRAM"),
        ("noparity_s4_fillmram_rf_p1_c0", None),
        ("noparity_s4_fillmram_rf",       None),
        ("noparity_s4_fillmram_p1_c0",    None),
        ("noparity_s4_fillmram",          "Static S4"),
    ]

    _MULT_COLORS = {2: "#1f77b4", 3: "#ff7f0e", 4: "#2ca02c", 5: "#d62728"}

    def _classify(var_str):
        for substr, label in _VARIANT_MAP:
            if substr in str(var_str):
                return label
        return None

    def _read_mult(tag):
        m = re.match(r"r?(\d+)x", str(tag))
        return int(m.group(1)) if m else None

    df = df.copy()
    df["design"] = df["variant"].apply(_classify)
    df = df[df["design"].notna()].copy()

    df["rd_mult"] = df["rw_tag"].apply(_read_mult)
    df["bench_short"] = df["bench"].apply(_short_bench)
    df["sz_mb"] = df["size_mb"]
    df["elapsed_fs"] = df["elapsed_ns"] * 1e6  # ns -> fs

    df = df[df["bench_short"].isin(_SHOW_BENCHES)].copy()

    df_base = df[df["design"] == "All-MRAM"].copy()
    df_sweep = df[df["rd_mult"].notna()].copy()

    rd_mults = sorted(df_sweep["rd_mult"].dropna().unique())

    # horizontal: 1 row, N cols
    ncols = len(_SHOW_BENCHES)
    fig, axes = plt.subplots(1, ncols, figsize=(5.2 * ncols, 4.0))
    if ncols == 1:
        axes = [axes]

    for ci, bench in enumerate(_SHOW_BENCHES):
        ax = axes[ci]
        b_base = df_base[df_base["bench_short"] == bench]
        b_sweep = df_sweep[df_sweep["bench_short"] == bench]

        x = np.arange(len(sizes))
        n_m = len(rd_mults)
        w = 0.7 / n_m if n_m > 0 else 0.7

        for mi, mult in enumerate(rd_mults):
            vals = []
            for sz in sizes:
                base_sub = b_base[
                    (b_base["sz_mb"] == sz) &
                    (b_base["rd_mult"] == mult)
                ]
                s4_sub = b_sweep[
                    (b_sweep["design"] == "Static S4") &
                    (b_sweep["rd_mult"] == mult) &
                    (b_sweep["sz_mb"] == sz)
                ]

                if base_sub.empty or s4_sub.empty:
                    vals.append(np.nan)
                else:
                    bt = base_sub["elapsed_fs"].values[0]
                    st = s4_sub["elapsed_fs"].values[0]
                    su = bt / st if st > 0 else np.nan
                    vals.append(su)
                    print(f"    {bench} {sz}MB {mult}x: speedup={su:.4f}")

            offset = (mi - n_m / 2) * w + w / 2
            ax.bar(
                x + offset, vals, w,
                color=_MULT_COLORS.get(mult, "#999"),
                edgecolor="white",
                label=f"{mult}x" if ci == 0 else None
            )

        ax.axhline(1.0, color="grey", ls="--", lw=1, zorder=0)
        ax.set_xticks(x)
        ax.set_xticklabels([f"{sz}MB" for sz in sizes], fontsize=10)
        ax.set_xlabel("Cache Size", fontsize=10)
        ax.set_title(bench, fontsize=11)
        ax.grid(axis="y", linestyle="--", alpha=0.3)
        ax.set_ylim(bottom=0.0)
        if ci == 0:
            ax.set_ylabel("Speedup vs All-MRAM", fontsize=10)

    h, l = axes[0].get_legend_handles_labels()
    plt.tight_layout()
    plt.subplots_adjust(top=0.82)
    fig.legend(
        h, l,
        loc="upper center", bbox_to_anchor=(0.5, 1.02),
        ncol=max(len(rd_mults), 1),
        frameon=False, fontsize=9,
        title="MRAM Read Multiplier", title_fontsize=9
    )

    plt.show()
    
plot_standalone_bar_crossnode(df_cross, sizes=sizes)
plot_standalone_scatter_hca(df_hybrid, sizes=sizes)

if not df_sweep.empty:
    plot_read_only_sweep(df_sweep, sizes=[16, 32, 128])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_path = "~/COSC_498/miniMXE/repro/agg/extra.csv"
df = pd.read_csv(csv_path)

name_map = {
    "perlbench": "perl",
    "gcc": "gcc",
    "mcf": "mcf",
    "omnetpp": "omnetpp",
    "xalancbmk": "xalan",
    "deepsjeng": "dsj",
    "leela": "leela",
    "xz": "xz",
    "exchange2": "exch",
    "fotonik3d": "foto",
}
df["bench_short"] = df["bench_short"].replace(name_map)

bench_order = ["perl", "gcc", "mcf", "omnetpp", "xalan", "dsj", "leela", "xz", "exch", "foto"]
xtick_labels = ["mean"] + bench_order

# -----------------------------
# Plot 1 only: Energy decomposition
# -----------------------------
e = df[df["table_group"] == "table8_energy"].copy()

configs = ["SRAM7", "MRAM14", "M+DVFS"]
raw = {cfg: {} for cfg in configs}

for cfg in configs:
    sub = e[e["variant"] == cfg].copy()
    for b in bench_order:
        r = sub[sub["bench_short"] == b]
        if len(r) == 0:
            continue
        r = r.iloc[0]
        raw[cfg][b] = [
            float(r["e_nc_dyn_j"]),
            float(r["e_nc_static_j"]),
            float(r["e_llc_dyn_j"]),
            float(r["e_llc_leak_j"]),
        ]

for cfg in configs:
    arr = np.array([raw[cfg][b] for b in bench_order], dtype=float)
    raw[cfg] = {"mean": arr.mean(axis=0), **raw[cfg]}

plot_order = ["mean"] + bench_order
vals = {cfg: np.array([raw[cfg][b] for b in plot_order], dtype=float) for cfg in configs}

baseline_total = vals["SRAM7"].sum(axis=1)
baseline_llc = vals["SRAM7"][:, 2] + vals["SRAM7"][:, 3]

norm_total = {}
norm_llc = {}

for cfg in configs:
    arr = vals[cfg]

    norm_total[cfg] = np.column_stack([
        arr[:, 0] / baseline_total,
        arr[:, 1] / baseline_total,
        arr[:, 2] / baseline_total,
        arr[:, 3] / baseline_total,
    ])

    norm_llc[cfg] = np.column_stack([
        np.zeros(len(plot_order)),
        np.zeros(len(plot_order)),
        np.divide(arr[:, 2], baseline_llc, out=np.zeros(len(plot_order)), where=baseline_llc > 0),
        np.divide(arr[:, 3], baseline_llc, out=np.zeros(len(plot_order)), where=baseline_llc > 0),
    ])

x = np.arange(len(plot_order))
w = 0.22
offsets = [-w, 0, w]

c_nc_dyn = "#4C72B0"
c_nc_static = "#C9D3DD"
c_llc_dyn = "#D8A03A"
c_llc_leak = "#D9534F"

plt.rcParams.update({
    "font.size": 14,
    "axes.labelsize": 16,
    "axes.titlesize": 16,
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
    "legend.fontsize": 13,
})

# Horizontal layout: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(17, 5), sharex=False)

# Left plot: total energy
ax = axes[0]
for i, cfg in enumerate(configs):
    xpos = x + offsets[i]
    arr = norm_total[cfg]

    ax.bar(
        xpos, arr[:, 0], width=w,
        color=c_nc_dyn,
        label="Non-cache dynamic" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )
    ax.bar(
        xpos, arr[:, 1], width=w, bottom=arr[:, 0],
        color=c_nc_static,
        label="Non-cache static" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )
    ax.bar(
        xpos, arr[:, 2], width=w, bottom=arr[:, 0] + arr[:, 1],
        color=c_llc_dyn,
        label="LLC dynamic" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )
    ax.bar(
        xpos, arr[:, 3], width=w, bottom=arr[:, 0] + arr[:, 1] + arr[:, 2],
        color=c_llc_leak,
        label="LLC leakage" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )

ax.set_ylabel("Normalized Total Energy")
ax.set_ylim(0, 1.12)
ax.set_xticks(x)
ax.set_xticklabels(xtick_labels, rotation=45, ha="right")
ax.grid(axis="y", alpha=0.25)

# Right plot: LLC energy
ax2 = axes[1]
for i, cfg in enumerate(configs):
    xpos = x + offsets[i]
    arr = norm_llc[cfg]

    ax2.bar(
        xpos, arr[:, 2], width=w,
        color=c_llc_dyn,
        label="LLC dynamic" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )
    ax2.bar(
        xpos, arr[:, 3], width=w, bottom=arr[:, 2],
        color=c_llc_leak,
        label="LLC leakage" if i == 0 else None,
        edgecolor="black", linewidth=0.8
    )

ax2.set_ylabel("Normalized LLC Energy")
ax2.set_ylim(0, 1.12)
ax2.set_xticks(x)
ax2.set_xticklabels(xtick_labels, rotation=45, ha="right")
ax2.grid(axis="y", alpha=0.25)

# One legend above both plots
handles, labels = ax.get_legend_handles_labels()
fig.legend(
    handles, labels,
    ncol=4,
    loc="upper center",
    bbox_to_anchor=(0.5, 1.05),
    frameon=False
)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csv_path = "~/COSC_498/miniMXE/repro/agg/extra.csv"
df = pd.read_csv(csv_path)

name_map = {
    "perlbench": "perl",
    "gcc": "gcc",
    "mcf": "mcf",
    "omnetpp": "omnetpp",
    "xalancbmk": "xalan",
    "deepsjeng": "dsj",
    "leela": "leela",
    "xz": "xz",
    "exchange2": "exch",
    "fotonik3d": "foto",
}
df["bench_short"] = df["bench_short"].replace(name_map)

bench_order = ["perl", "gcc", "mcf", "omnetpp", "xalan", "dsj", "leela", "xz", "exch", "foto"]

plt.rcParams.update({
    "font.size": 14,
    "axes.labelsize": 16,
    "axes.titlesize": 16,
    "xtick.labelsize": 13,
    "ytick.labelsize": 13,
    "legend.fontsize": 13,
})

# -----------------------------
# Data for Plot 1: Read latency sweep
# -----------------------------
r = df[df["table_group"] == "table10_readlat"].copy()
r = r[(r["variant"] == "M+DVFS") & (r["n_tag"] == "n1")].copy()
r["sweep_value"] = pd.to_numeric(r["sweep_value"], errors="coerce")

lat_mults = [2, 3, 4, 5]

# -----------------------------
# Data for Plot 2: Cap +/- MAE
# -----------------------------
c = df[df["table_group"] == "table12_cap_mae"].copy()
c = c[c["n_tag"] == "n1"].copy()

# -----------------------------
# Combined figure: 2 plots side by side
# -----------------------------
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 3.6), sharey=True)

# ---- Left: Read latency sweep ----
x1 = np.arange(len(bench_order))
n_bars = len(lat_mults)
w1 = 0.8 / n_bars

for bi, mult in enumerate(lat_mults):
    vals = []
    for b in bench_order:
        rr = r[(r["bench_short"] == b) & (r["sweep_value"] == float(mult))]
        if len(rr) == 0:
            vals.append(np.nan)
        else:
            vals.append(1.0 + float(rr.iloc[0]["speedup_pct_vs_sram7"]) / 100.0)

    ax1.bar(
        x1 + (bi - (n_bars - 1) / 2.0) * w1,
        vals,
        w1,
        label=f"{mult}x",
        edgecolor="black",
        linewidth=0.8,
    )

ax1.axhline(1.0, color="gray", linestyle="--")
ax1.set_xticks(x1)
ax1.set_xticklabels(bench_order, rotation=45, ha="right")
ax1.set_ylabel("Speedup vs SRAM7")
ax1.set_title("Read Latency Sweep (n=1, 16MB)")
ax1.grid(axis="y", alpha=0.25)
ax1.legend()

# ---- Right: Cap +/- MAE ----
x2 = np.arange(len(bench_order))
w2 = 0.35

minus_vals = []
plus_vals = []

for b in bench_order:
    rm = c[(c["bench_short"] == b) & (c["variant"] == "cap_minus_mae")]
    rp = c[(c["bench_short"] == b) & (c["variant"] == "cap_plus_mae")]

    minus_vals.append(1.0 + float(rm.iloc[0]["speedup_pct_vs_sram7"]) / 100.0 if len(rm) else np.nan)
    plus_vals.append(1.0 + float(rp.iloc[0]["speedup_pct_vs_sram7"]) / 100.0 if len(rp) else np.nan)

ax2.bar(x2 - w2/2, minus_vals, w2, label="Cap-MAE", edgecolor="black", linewidth=0.8)
ax2.bar(x2 + w2/2, plus_vals,  w2, label="Cap+MAE", edgecolor="black", linewidth=0.8)

ax2.axhline(1.0, color="gray", linestyle="--")
ax2.set_xticks(x2)
ax2.set_xticklabels(bench_order, rotation=45, ha="right")
ax2.set_title("Cap +/- MAE (n=1, 16MB)")
ax2.grid(axis="y", alpha=0.25)
ax2.legend()

plt.tight_layout()
plt.show()

# -----------------------------
# Optional raw-value prints
# -----------------------------
print("\n=== Read latency raw values (ratio = 1 + speedup/100) ===")
rl_rows = []
for b in bench_order:
    row = {"bench": b}
    for mult in lat_mults:
        rr = r[(r["bench_short"] == b) & (r["sweep_value"] == float(mult))]
        row[f"x{mult}_speedup_pct"] = float(rr.iloc[0]["speedup_pct_vs_sram7"]) if len(rr) else np.nan
        row[f"x{mult}_ratio"] = 1.0 + row[f"x{mult}_speedup_pct"] / 100.0 if len(rr) else np.nan
    rl_rows.append(row)

rl_df = pd.DataFrame(rl_rows)
print(rl_df.to_string(index=False, float_format=lambda x: f"{x:.6f}"))

print("\n=== Cap +/- MAE raw values (ratio = 1 + speedup/100) ===")
cap_rows = []
for b, mv, pv in zip(bench_order, minus_vals, plus_vals):
    rm = c[(c["bench_short"] == b) & (c["variant"] == "cap_minus_mae")]
    rp = c[(c["bench_short"] == b) & (c["variant"] == "cap_plus_mae")]

    m_pct = float(rm.iloc[0]["speedup_pct_vs_sram7"]) if len(rm) else np.nan
    p_pct = float(rp.iloc[0]["speedup_pct_vs_sram7"]) if len(rp) else np.nan

    cap_rows.append({
        "bench": b,
        "cap_minus_speedup_pct": m_pct,
        "cap_minus_ratio": mv,
        "cap_plus_speedup_pct": p_pct,
        "cap_plus_ratio": pv,
    })

cap_df = pd.DataFrame(cap_rows)
print(cap_df.to_string(index=False, float_format=lambda x: f"{x:.6f}"))
```

```python
#!/usr/bin/env python3
import csv
import math
import os
import re
from typing import Dict, List, Tuple, Optional

import pandas as pd
import matplotlib.pyplot as plt

OURS_CSV = os.path.expanduser("/home/skataoka26/COSC_498/miniMXE/repro/agg/dvfs.csv")
GLOBAL_CSV = os.path.expanduser("/home/skataoka26/COSC_498/miniMXE/repro/agg/global_dvfs.csv")
SELECTIVE_NOH_CSV = os.path.expanduser("/home/skataoka26/COSC_498/miniMXE/repro/agg/selective_dvfs.csv")

SIZE_MB = 16

# Exclude the one incomplete workload row from all comparisons
EXCLUDE_KEYS = {
    ("505.mcf_r+500.perlbench_r+648.exchange2_s+649.fotonik3d_s", "n4"),
}

label_map = {
    ("500.perlbench_r", "n1"): "perl",
    ("502.gcc_r", "n1"): "gcc",
    ("505.mcf_r", "n1"): "mcf",
    ("520.omnetpp_r", "n1"): "omnetpp",
    ("523.xalancbmk_r", "n1"): "xalan",
    ("531.deepsjeng_r", "n1"): "dsj",
    ("541.leela_r", "n1"): "leela",
    ("557.xz_r", "n1"): "xz",
    ("648.exchange2_s", "n1"): "exch",
    ("649.fotonik3d_s", "n1"): "foto",

    ("502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n4"): "gcc×4",
    ("505.mcf_r+500.perlbench_r+648.exchange2_s+649.fotonik3d_s", "n4"): "mcf+perl+exch+foto",
    ("505.mcf_r+505.mcf_r+502.gcc_r+502.gcc_r", "n4"): "mcf×2+gcc×2",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r", "n4"): "mcf×4",
    ("523.xalancbmk_r+523.xalancbmk_r+502.gcc_r+502.gcc_r", "n4"): "xalan×2+gcc×2",

    ("502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "gcc×8",
    ("505.mcf_r+505.mcf_r+500.perlbench_r+500.perlbench_r+648.exchange2_s+648.exchange2_s+649.fotonik3d_s+649.fotonik3d_s", "n8"): "mcf×2+perl×2+exch×2+foto×2",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "mcf×4+gcc×4",
    ("505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r", "n8"): "mcf×8",
    ("523.xalancbmk_r+523.xalancbmk_r+523.xalancbmk_r+523.xalancbmk_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r", "n8"): "xalan×4+gcc×4",
}

POLICIES = [
    ("Global DVFS", GLOBAL_CSV),
    ("Selective w/o heuristics", SELECTIVE_NOH_CSV),
    ("Selective + heuristics (ours)", OURS_CSV),
]

warnings = []


def policy_slug(policy_name: str) -> str:
    return (
        policy_name.lower()
        .replace(" + ", "_")
        .replace(" ", "_")
        .replace("/", "over")
        .replace("(", "")
        .replace(")", "")
        .replace("-", "_")
    )


def parse_vec(s: str) -> List[float]:
    if not s or str(s).strip() == "":
        return []
    return [float(x) for x in s.split("|") if x.strip()]


def total_n_from_label(label: str) -> int:
    total = 0
    for token in label.split("+"):
        m = re.search(r"×(\d+)", token)
        total += int(m.group(1)) if m else 1
    return total


def shorten_label(label: str) -> str:
    return label.replace("omnetpp", "omnet")


def smart_multiline_label(label: str, max_parts_per_line: int = 2) -> str:
    parts = shorten_label(label).split("+")
    lines = []
    for i in range(0, len(parts), max_parts_per_line):
        chunk = parts[i:i + max_parts_per_line]
        line = "+".join(chunk)
        if i > 0:
            line = "+" + line
        lines.append(line)
    return "\n".join(lines)


def load_policy_rows(csv_path: str, size_mb: int) -> Dict[Tuple[str, str], Dict[str, str]]:
    out = {}
    with open(csv_path, newline="") as f:
        reader = csv.DictReader(f)
        for row in reader:
            if row["size_mb"] != str(size_mb):
                continue
            key = (row["workload"], row["n_tag"])
            if key not in label_map:
                continue
            out[key] = row
    return out


def common_keys(dicts: List[Dict[Tuple[str, str], Dict[str, str]]]) -> List[Tuple[str, str]]:
    keys = set(dicts[0].keys())
    for d in dicts[1:]:
        keys &= set(d.keys())
    keys = [k for k in keys if k not in EXCLUDE_KEYS]
    return sorted(keys, key=lambda k: (total_n_from_label(label_map[k]), label_map[k]))


def get_ws_vs_sram7(row: Dict[str, str]) -> float:
    if row["n_tag"] == "n1":
        v = row.get("speedup_vs_sram7_ratio", "").strip()
        if v != "":
            return float(v)

        dvfs = row.get("dvfs_makespan_s", "").strip()
        sram = row.get("sram7_makespan_s", "").strip()
        if dvfs != "" and sram != "":
            return float(sram) / float(dvfs)

        raise ValueError(
            f"Missing n1 speedup for workload={row['workload']} n_tag={row['n_tag']}"
        )

    v = row.get("wsn_vs_sram7", "").strip()
    if v != "":
        return float(v)

    terms = parse_vec(row.get("ws_terms_vs_sram7", ""))
    if terms:
        return sum(terms)

    raise ValueError(
        f"Missing weighted speedup for workload={row['workload']} n_tag={row['n_tag']}"
    )


def align_baseline_and_policy_times(
    row: Dict[str, str],
    sram_times: List[float],
    policy_times: List[float],
) -> Optional[Tuple[List[float], List[float]]]:
    if len(sram_times) == len(policy_times):
        return sram_times, policy_times

    if len(policy_times) == 0:
        return None

    if row["n_tag"] == "n1" and len(sram_times) == 1 and len(policy_times) >= 1:
        return [sram_times[0]], [policy_times[0]]

    if len(policy_times) == len(sram_times) + 1:
        return sram_times, policy_times[:len(sram_times)]

    return None


policy_rows = [load_policy_rows(path, SIZE_MB) for _, path in POLICIES]
keys = common_keys(policy_rows)

if not keys:
    raise RuntimeError("No common workloads found across the three CSV files for SIZE_MB after exclusions.")

summary_records = []

for key in keys:
    workload, n_tag = key
    label = label_map[key]
    n_total = total_n_from_label(label)

    base_row = policy_rows[0][key]
    raw_sram_times = parse_vec(base_row["sram7_times_s"])

    rec = {
        "workload": workload,
        "n_tag": n_tag,
        "N": n_total,
        "label": label,
        "label_plot": smart_multiline_label(label),
        "raw_sram7_times_s": "|".join(f"{x:.9f}" for x in raw_sram_times),
    }

    for policy_name, rows in zip([p[0] for p in POLICIES], policy_rows):
        row = rows[key]
        slug = policy_slug(policy_name)

        ws = get_ws_vs_sram7(row)
        raw_policy_times = parse_vec(row["dvfs_times_s"])
        aligned = align_baseline_and_policy_times(row, raw_sram_times, raw_policy_times)

        rec[f"{slug}_ws_vs_sram7"] = ws
        rec[f"{slug}_raw_times_s"] = "|".join(f"{x:.9f}" for x in raw_policy_times)

        if aligned is None:
            rec[f"{slug}_times_s"] = ""
            rec[f"{slug}_speedups"] = ""
            warnings.append(
                f"Skipping per-core data for {policy_name} on {label} "
                f"(workload={workload}, n_tag={n_tag}, len(policy)={len(raw_policy_times)}, len(sram)={len(raw_sram_times)})"
            )
        else:
            aligned_sram_times, aligned_policy_times = aligned
            per_core_speedups = [
                (b / p) if p > 0 else float("nan")
                for b, p in zip(aligned_sram_times, aligned_policy_times)
            ]
            rec[f"{slug}_times_s"] = "|".join(f"{x:.9f}" for x in aligned_policy_times)
            rec[f"{slug}_speedups"] = "|".join(
                "nan" if math.isnan(x) else f"{x:.9f}" for x in per_core_speedups
            )

    summary_records.append(rec)

summary_df = pd.DataFrame(summary_records)

ours_slug = policy_slug("Selective + heuristics (ours)")
sort_col = f"{ours_slug}_ws_vs_sram7"

groups = []
boundaries = []
cursor = 0

for n in [1, 4, 8]:
    g = (
        summary_df[summary_df["N"] == n]
        .sort_values(sort_col, ascending=False)
        .reset_index(drop=True)
    )
    if len(g) == 0:
        continue
    groups.append(g)
    cursor += len(g)
    boundaries.append(cursor - 0.5)

plot_df = pd.concat(groups, ignore_index=True)
x = list(range(len(plot_df)))

fig, ax = plt.subplots(figsize=(14, 4.8))

for policy_name, _ in POLICIES:
    slug = policy_slug(policy_name)
    ax.plot(
        x,
        plot_df[f"{slug}_ws_vs_sram7"],
        marker="o",
        linestyle="-",
        label=policy_name,
    )

ax.axhline(1.0, linestyle="--", linewidth=1)

for boundary in boundaries[:-1]:
    ax.axvline(boundary, linestyle=":", linewidth=1)

ax.set_xticks(x)
ax.set_xticklabels(plot_df["label_plot"], rotation=45, ha="right")
ax.set_ylabel("Speedup vs SRAM7\n(n1: scalar speedup, n4/n8: weighted speedup)")
ax.set_xlabel("Workload / mix")
ax.grid(axis="y", alpha=0.25)
ax.legend(frameon=False, ncol=3, loc="upper right")

group_sizes = [len(g) for g in groups]
starts = []
running = 0
for sz in group_sizes:
    starts.append(running)
    running += sz

centers = [st + (sz - 1) / 2 for st, sz in zip(starts, group_sizes)]
group_labels = []
if len(group_sizes) >= 1:
    group_labels.append("N=1")
if len(group_sizes) >= 2:
    group_labels.append("N=4")
if len(group_sizes) >= 3:
    group_labels.append("N=8")

for center, label in zip(centers, group_labels):
    ax.text(center, 1.01, label, transform=ax.get_xaxis_transform(),
            ha="center", va="bottom")

plt.tight_layout()
plt.show()

elapsed_records = []
speedup_records = []

for key in keys:
    label = label_map[key]
    n_total = total_n_from_label(label)
    raw_sram_times = parse_vec(policy_rows[0][key]["sram7_times_s"])

    sram_added_for_this_key = False

    for policy_name, rows in zip([p[0] for p in POLICIES], policy_rows):
        row = rows[key]
        raw_policy_times = parse_vec(row["dvfs_times_s"])
        aligned = align_baseline_and_policy_times(row, raw_sram_times, raw_policy_times)

        if aligned is None:
            continue

        aligned_sram_times, aligned_policy_times = aligned

        if not sram_added_for_this_key:
            for core_id, sram_t in enumerate(aligned_sram_times):
                elapsed_records.append({
                    "workload": key[0],
                    "n_tag": key[1],
                    "N": n_total,
                    "label": label,
                    "core": core_id,
                    "policy": "SRAM7",
                    "elapsed_s": sram_t,
                })
            sram_added_for_this_key = True

        for core_id, (sram_t, pol_t) in enumerate(zip(aligned_sram_times, aligned_policy_times)):
            elapsed_records.append({
                "workload": key[0],
                "n_tag": key[1],
                "N": n_total,
                "label": label,
                "core": core_id,
                "policy": policy_name,
                "elapsed_s": pol_t,
            })
            speedup_records.append({
                "workload": key[0],
                "n_tag": key[1],
                "N": n_total,
                "label": label,
                "core": core_id,
                "policy": policy_name,
                "speedup_vs_sram7": (sram_t / pol_t) if pol_t > 0 else float("nan"),
            })

elapsed_df = pd.DataFrame(elapsed_records)
speedup_df = pd.DataFrame(speedup_records)

LEFT_WORKLOAD = "523.xalancbmk_r+523.xalancbmk_r+502.gcc_r+502.gcc_r"
RIGHT_WORKLOAD = "505.mcf_r+505.mcf_r+505.mcf_r+505.mcf_r+502.gcc_r+502.gcc_r+502.gcc_r+502.gcc_r"

selected_keys = []
for key in keys:
    if key[0] in {LEFT_WORKLOAD, RIGHT_WORKLOAD}:
        selected_keys.append(key)

if selected_keys:
    n_panels = len(selected_keys)

    fig, axes = plt.subplots(1, n_panels, figsize=(6.6 * n_panels, 4.8), squeeze=False)
    axes = axes[0]

    for ax, key in zip(axes, selected_keys):
        label = label_map[key]
        raw_sram_times = parse_vec(policy_rows[0][key]["sram7_times_s"])

        aligned_sram_ref = None
        for _policy_name, rows in zip([p[0] for p in POLICIES], policy_rows):
            row = rows[key]
            aligned = align_baseline_and_policy_times(row, raw_sram_times, parse_vec(row["dvfs_times_s"]))
            if aligned is not None:
                aligned_sram_ref, _ = aligned
                break

        if aligned_sram_ref is None:
            ax.set_title(f"{label}\n(no per-core data)")
            ax.set_xlabel("Core")
            ax.set_ylabel("Elapsed time (s)")
            continue

        xs = list(range(len(aligned_sram_ref)))
        ax.plot(xs, aligned_sram_ref, marker="o", linestyle="-", label="SRAM7")

        for policy_name, rows in zip([p[0] for p in POLICIES], policy_rows):
            row = rows[key]
            aligned = align_baseline_and_policy_times(row, raw_sram_times, parse_vec(row["dvfs_times_s"]))
            if aligned is None:
                continue
            _, aligned_policy_times = aligned
            xs_pol = list(range(len(aligned_policy_times)))
            ax.plot(xs_pol, aligned_policy_times, marker="o", linestyle="-", label=policy_name)

        ax.set_title(label)
        ax.set_xlabel("Core")
        ax.set_ylabel("Elapsed time (s)")
        ax.set_xticks(xs)
        ax.grid(axis="y", alpha=0.25)

    handles, labels = axes[0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="upper center", ncol=4, frameon=False)
    plt.tight_layout(rect=[0, 0, 1, 0.92])
    plt.show()

    fig, axes = plt.subplots(1, n_panels, figsize=(6.6 * n_panels, 4.8), squeeze=False)
    axes = axes[0]

    for ax, key in zip(axes, selected_keys):
        label = label_map[key]
        raw_sram_times = parse_vec(policy_rows[0][key]["sram7_times_s"])
        last_xs = None

        for policy_name, rows in zip([p[0] for p in POLICIES], policy_rows):
            row = rows[key]
            aligned = align_baseline_and_policy_times(row, raw_sram_times, parse_vec(row["dvfs_times_s"]))
            if aligned is None:
                continue

            aligned_sram_times, aligned_policy_times = aligned
            sp = [(b / p) if p > 0 else float("nan") for b, p in zip(aligned_sram_times, aligned_policy_times)]
            xs = list(range(len(sp)))
            last_xs = xs
            ax.plot(xs, sp, marker="o", linestyle="-", label=policy_name)

        ax.axhline(1.0, linestyle="--", linewidth=1)
        ax.set_title(label)
        ax.set_xlabel("Core")
        ax.set_ylabel("Per-core speedup vs SRAM7")
        if last_xs is not None:
            ax.set_xticks(last_xs)
        ax.grid(axis="y", alpha=0.25)

    handles, labels = axes[0].get_legend_handles_labels()
    if handles:
        fig.legend(handles, labels, loc="upper center", ncol=3, frameon=False)
    plt.tight_layout(rect=[0, 0, 1, 0.92])
    plt.show()

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)

print("\nExcluded workload keys:")
for k in sorted(EXCLUDE_KEYS):
    print(" ", k)

print("\nWeighted-speedup summary:")
cols = ["label", "N"]
for policy_name, _ in POLICIES:
    cols.append(f"{policy_slug(policy_name)}_ws_vs_sram7")
print(plot_df[cols].to_string(index=False))

print("\nPer-core elapsed-time table:")
print(elapsed_df.to_string(index=False))

print("\nPer-core speedup table:")
print(speedup_df.to_string(index=False))

if warnings:
    print("\nWarnings:")
    for w in warnings:
        print(" -", w)
```

```python
#!/usr/bin/env python3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CSV = "/home/skataoka26/COSC_498/miniMXE/repro/agg/workload.csv"

NAME_MAP = {
    "505.mcf_r": "mcf",
    "523.xalancbmk_r": "xalancbmk",
    "500.perlbench_r": "perlbench",
    "520.omnetpp_r": "omnetpp",
    "502.gcc_r": "gcc",
    "557.xz_r": "xz",
    "531.deepsjeng_r": "deepsjeng",
    "541.leela_r": "leela",
    "649.fotonik3d_s": "fotonik3d",
    "648.exchange2_s": "exchange2",
}

BASE_ORDER = [
    "mcf", "xalancbmk", "perlbench", "omnetpp", "gcc",
    "xz", "deepsjeng", "leela", "fotonik3d", "exchange2"
]

df = pd.read_csv(CSV)
df["short"] = df["workload"].map(NAME_MAP)

# -------------------------------------------------
# Helpers
# -------------------------------------------------
def add_write_frac(d):
    d = d.copy()
    total_hits = d["l3_read_hits"] + d["l3_write_hits"]
    d["write_hit_frac"] = np.where(total_hits > 0, d["l3_write_hits"] / total_hits, 0.0)
    return d

def make_taxonomy(base_df):
    """
    Rule-based buckets using one baseline design point.
    Adjust thresholds after seeing the data distribution.
    """
    d = add_write_frac(base_df).copy()

    total_hits = d["l3_read_hits"] + d["l3_write_hits"]

    buckets = []
    for _, r in d.iterrows():
        if r["write_hit_frac"] >= 0.35:
            bucket = "write-active"
        elif r["mem_cpi_frac"] <= 0.10 and total_hits.loc[r.name] < total_hits.median():
            bucket = "LLC-light"
        elif r["l3_miss_rate"] >= 0.20 or r["long_lat_frac"] >= 0.35:
            bucket = "miss-dominated"
        else:
            bucket = "hit-intensive"
        buckets.append(bucket)

    d["bucket"] = buckets
    return d[[
        "short", "throughput_inst_ns", "l3_miss_rate", "long_lat_frac",
        "mem_cpi_frac", "write_hit_frac", "bucket"
    ]].sort_values("short")

def compare_points(df, base_cfg, base_sz, target_cfgs, n_tag="n1"):
    """
    Build a long comparison table:
      base point = (base_cfg, base_sz)
      targets = list of (cfg, sz, label)
    """
    cols = [
        "short", "config", "size_mb", "makespan_s", "l3_miss_rate",
        "long_lat_frac", "mem_cpi_frac", "throughput_inst_ns"
    ]
    sub = df[(df["n_tag"] == n_tag) & (df["short"].notna())][cols].copy()

    base = sub[
        (sub["config"] == base_cfg) &
        (sub["size_mb"] == base_sz)
    ].copy()

    base = base.rename(columns={
        "makespan_s": "rt_base",
        "l3_miss_rate": "miss_base",
        "long_lat_frac": "longlat_base",
        "mem_cpi_frac": "memcpi_base",
        "throughput_inst_ns": "ipn_base",
    })[["short", "rt_base", "miss_base", "longlat_base", "memcpi_base", "ipn_base"]]

    outs = []
    for cfg, sz, label in target_cfgs:
        tgt = sub[
            (sub["config"] == cfg) &
            (sub["size_mb"] == sz)
        ].copy()

        tgt = tgt.rename(columns={
            "makespan_s": "rt_tgt",
            "l3_miss_rate": "miss_tgt",
            "long_lat_frac": "longlat_tgt",
            "mem_cpi_frac": "memcpi_tgt",
            "throughput_inst_ns": "ipn_tgt",
        })[["short", "rt_tgt", "miss_tgt", "longlat_tgt", "memcpi_tgt", "ipn_tgt"]]

        m = base.merge(tgt, on="short", how="inner")
        m["target_label"] = label

        m["miss_reduction"] = np.where(
            m["miss_base"] > 0,
            (m["miss_base"] - m["miss_tgt"]) / m["miss_base"],
            np.nan
        )
        m["runtime_change"] = np.where(
            m["rt_base"] > 0,
            (m["rt_base"] - m["rt_tgt"]) / m["rt_base"],
            np.nan
        )
        m["longlat_change"] = m["longlat_tgt"] - m["longlat_base"]
        m["memcpi_change"] = m["memcpi_tgt"] - m["memcpi_base"]

        outs.append(m)

    return pd.concat(outs, ignore_index=True)

def print_summary_table(comp):
    out = comp.copy()
    out["miss_reduction"] = out["miss_reduction"].map(lambda x: f"{100*x:.1f}%")
    out["runtime_change"] = out["runtime_change"].map(lambda x: f"{100*x:.1f}%")
    out["longlat_change"] = out["longlat_change"].map(lambda x: f"{100*x:+.1f} pp")
    out["memcpi_change"] = out["memcpi_change"].map(lambda x: f"{100*x:+.1f} pp")
    print(out[[
        "short", "target_label", "miss_reduction",
        "runtime_change", "longlat_change", "memcpi_change"
    ]].sort_values(["target_label", "short"]).to_string(index=False))

def plot_miss_vs_runtime(comp, title):
    fig, ax = plt.subplots(figsize=(7, 5))

    marker_map = {
        "32MB MRAM14": "o",
        "128MB MRAM14": "s",
        "32MB SRAM7": "^",
        "128MB SRAM7": "D",
    }

    for label, g in comp.groupby("target_label"):
        ax.scatter(
            g["miss_reduction"],
            g["runtime_change"],
            s=70,
            marker=marker_map.get(label, "o"),
            label=label,
            alpha=0.9
        )
        for _, r in g.iterrows():
            ax.annotate(
                r["short"],
                (r["miss_reduction"], r["runtime_change"]),
                xytext=(4, 4),
                textcoords="offset points",
                fontsize=8
            )

    ax.axhline(0, color="gray", linestyle="--", linewidth=1)
    ax.axvline(0, color="gray", linestyle="--", linewidth=1)
    ax.set_xlabel("Relative miss-rate reduction")
    ax.set_ylabel("Relative runtime improvement")
    ax.set_title(title)
    ax.grid(alpha=0.3, linestyle="--")
    ax.legend(frameon=False)
    plt.tight_layout()
    plt.show()

# -------------------------------------------------
# 1) Taxonomy from one baseline point
# -------------------------------------------------
baseline_tax = df[
    (df["n_tag"] == "n1") &
    (df["size_mb"] == 16) &
    (df["config"] == "MRAM14") &
    (df["short"].notna())
].copy()

tax = make_taxonomy(baseline_tax)
tax["Workload"] = pd.Categorical(tax["short"], categories=BASE_ORDER, ordered=True)
tax = tax.sort_values("Workload")

tax_out = tax.copy()
tax_out["throughput_inst_ns"] = tax_out["throughput_inst_ns"].map(lambda x: f"{x:.2f}")
tax_out["l3_miss_rate"] = tax_out["l3_miss_rate"].map(lambda x: f"{100*x:.1f}%")
tax_out["long_lat_frac"] = tax_out["long_lat_frac"].map(lambda x: f"{100*x:.1f}%")
tax_out["mem_cpi_frac"] = tax_out["mem_cpi_frac"].map(lambda x: f"{100*x:.1f}%")
tax_out["write_hit_frac"] = tax_out["write_hit_frac"].map(lambda x: f"{100*x:.1f}%")

print("\nTaxonomy table (baseline: 16MB MRAM14)\n")
print(tax_out[[
    "short", "throughput_inst_ns", "l3_miss_rate",
    "long_lat_frac", "mem_cpi_frac", "write_hit_frac", "bucket"
]].to_string(index=False))

# -------------------------------------------------
# 2) Capacity scaling inside MRAM only
# -------------------------------------------------
comp_mram = compare_points(
    df,
    base_cfg="MRAM14",
    base_sz=16,
    target_cfgs=[
        ("MRAM14", 32, "32MB MRAM14"),
        ("MRAM14", 128, "128MB MRAM14"),
    ],
    n_tag="n1"
)

print("\nCapacity scaling summary (base = 16MB MRAM14)\n")
print_summary_table(comp_mram)

plot_miss_vs_runtime(
    comp_mram,
    "Miss-rate reduction vs runtime improvement\n(base = 16MB MRAM14)"
)
```

```python
import os
import re
import sqlite3
import pandas as pd

BASE = "/home/skataoka26/COSC_498/miniMXE/repro"
TUNING_ROOT = "/home/skataoka26/COSC_498/miniMXE/repro/dvfs/15_tuning/runs"

BASELINE_RUN_ROOT = os.path.join(BASE, "dvfs", "12_baseline_run", "runs")
MAIN_DVFS_ROOT = os.path.join(BASE, "dvfs", "1_main_dvfs", "runs")
HCA_ROOT = os.path.join(BASE, "hca", "hca_sunnycove")


def _get_stat(db_path, obj, metric, prefix="roi-end", core=0):
    try:
        conn = sqlite3.connect(f"file:{db_path}?mode=ro", uri=True)
        c = conn.cursor()
        c.execute(
            '''SELECT v.value FROM "values" v
               JOIN names n ON v.nameid = n.nameid
               JOIN prefixes p ON v.prefixid = p.prefixid
               WHERE n.objectname=? AND n.metricname=? AND p.prefixname=? AND v.core=?''',
            (obj, metric, prefix, core),
        )
        row = c.fetchone()
        conn.close()
        return float(row[0]) if row else None
    except Exception:
        return None


def get_delta(db_path, obj, metric, core=0):
    end = _get_stat(db_path, obj, metric, "roi-end", core)
    begin = _get_stat(db_path, obj, metric, "roi-begin", core)
    if end is not None and begin is not None:
        return end - begin
    return end


def get_num_cores(db_path):
    try:
        conn = sqlite3.connect(f"file:{db_path}?mode=ro", uri=True)
        c = conn.cursor()
        c.execute('SELECT MAX(v.core) FROM "values" v')
        row = c.fetchone()
        conn.close()
        return int(row[0]) + 1 if row and row[0] is not None else 1
    except Exception:
        return 1


def _read_sim_n(run_dir):
    try:
        with open(os.path.join(run_dir, "run.yaml")) as f:
            for line in f:
                line = line.strip()
                if line.startswith("sim_n:"):
                    return int(line.split(":")[1].strip())
    except Exception:
        pass
    return None


def _parse_sim_out_times(path):
    begins, ends, elapsed = [], [], []
    try:
        with open(path) as f:
            for line in f:
                s = line.strip()
                if s.startswith("performance_model.elapsed_time_begin"):
                    begins = [float(x) for x in s.split("=")[1].split(",") if x.strip()]
                elif s.startswith("performance_model.elapsed_time_end"):
                    ends = [float(x) for x in s.split("=")[1].split(",") if x.strip()]
                elif s.startswith("performance_model.elapsed_time"):
                    if "begin" not in s and "end" not in s:
                        elapsed = [float(x) for x in s.split("=")[1].split(",") if x.strip()]
    except Exception:
        pass

    if begins and ends and len(begins) == len(ends):
        return [(e - b) * 1e-15 for b, e in zip(begins, ends)]
    if elapsed:
        return [e * 1e-15 for e in elapsed]
    return None


def get_times_from_dir(run_dir):
    if not run_dir or not os.path.isdir(run_dir):
        return None

    sim_n = _read_sim_n(run_dir)

    sim_out = os.path.join(run_dir, "sim.out")
    times = _parse_sim_out_times(sim_out)
    if times:
        if sim_n and len(times) > sim_n:
            times = times[:sim_n]
        return times

    db_path = os.path.join(run_dir, "sim.stats.sqlite3")
    if not os.path.exists(db_path):
        return None

    n_cores = sim_n or get_num_cores(db_path)
    result = []
    for core in range(n_cores):
        t = get_delta(db_path, "thread", "elapsed_time", core)
        if t is None or t <= 0:
            t = get_delta(db_path, "performance_model", "elapsed_time", core)
        result.append(t * 1e-15 if t is not None and t > 0 else None)

    if any(v is not None and v > 0 for v in result):
        return result
    return None


def active_times_from_dir(run_dir):
    ts = get_times_from_dir(run_dir)
    if not ts:
        return None
    vals = [t for t in ts if t is not None and t > 0]
    return vals if vals else None


def makespan_s(run_dir):
    ts = active_times_from_dir(run_dir)
    if not ts:
        return None
    return max(ts)


def single_speedup_ratio(base_dir, variant_dir):
    b = makespan_s(base_dir)
    v = makespan_s(variant_dir)
    if b is None or v is None or v <= 0:
        return None
    return b / v


def _has_results(run_dir):
    if not run_dir or not os.path.isdir(run_dir):
        return False
    return (
        os.path.exists(os.path.join(run_dir, "sim.out")) or
        os.path.exists(os.path.join(run_dir, "sim.stats.sqlite3"))
    )


def find_main_baseline_dir(bench, n_tag, size_mb, leaf_exact):
    sz_dir = os.path.join(MAIN_DVFS_ROOT, bench, n_tag, f"l3_{size_mb}MB")
    if not os.path.isdir(sz_dir):
        return None
    exact = os.path.join(sz_dir, leaf_exact)
    if os.path.isdir(exact):
        return exact
    return None


def find_baseline_run_dir(bench, n_tag, size_mb, leaf_exact):
    d = os.path.join(BASELINE_RUN_ROOT, bench, n_tag, f"l3_{size_mb}MB", leaf_exact)
    return d if os.path.isdir(d) else None


def find_hca_baseline_dir_n1(bench, size_mb, leaf_exact):
    bench_dot = bench
    bench_us = bench.replace(".", "_", 1)
    sz_tag = f"sz{size_mb}M"
    campaigns = [
        "1_baselines",
        "2_cross_node/mram14",
        "2_cross_node/mram32",
        "3_static_hca",
    ]
    for campaign in campaigns:
        runs_dir = os.path.join(HCA_ROOT, campaign, "runs")
        if not os.path.isdir(runs_dir):
            continue
        for root, _, _ in os.walk(runs_dir):
            parts = root.split(os.sep)
            if parts[-1] != leaf_exact:
                continue
            if sz_tag not in parts:
                continue
            if not any(p.startswith(bench_dot) or p.startswith(bench_us) for p in parts):
                continue
            return root
    return None


def find_sram7_baseline(bench, size_mb):
    d = find_baseline_run_dir(bench, "n1", size_mb, "baseline_sram_only_sram7")
    if d and _has_results(d):
        return d

    d = find_main_baseline_dir(bench, "n1", size_mb, "baseline_sram_only_sram7")
    if d and _has_results(d):
        return d

    d = find_hca_baseline_dir_n1(bench, size_mb, "baseline_sram_only_sram7")
    if d and _has_results(d):
        return d

    return None


_LC_CHANGE_RE = re.compile(
    r"\[LC\] DVFS Change \[PLM\]: "
    r"P_est=([\d.]+)W "
    r"\(llc_leak=([\d.]+)W P_nocache=([\d.]+)W\) "
    r"Target=([\d.]+)W "
    r"f_lookup=([\d.]+)GHz\(\w+\) "
    r"u_sum=([\d.]+) "
    r"ipc=([\d.]+) "
    r"u_sum_x_ipc=([\d.]+) "
    r"boosted=(\d+)/(\d+) "
    r"f\[min/avg/max\]=\[([\d.]+)/([\d.]+)/([\d.]+)\] GHz"
)


def parse_sniper_log(log_path):
    intervals = []
    if not os.path.exists(log_path):
        return intervals

    try:
        with open(log_path) as f:
            for line in f:
                m = _LC_CHANGE_RE.search(line)
                if m:
                    intervals.append({
                        "P_est": float(m.group(1)),
                        "Target": float(m.group(4)),
                        "f_max": float(m.group(13)),
                    })
    except Exception:
        return []

    return intervals


def metrics_from_run(run_dir):
    out = {
        "mean_power_w": None,
        "frac_over_pcap": None,
        "num_transitions": None,
        "avg_freq_ghz": None,
        "max_freq_ghz": None,
    }

    intervals = parse_sniper_log(os.path.join(run_dir, "sniper.log"))
    if not intervals:
        return out

    n = len(intervals)
    out["mean_power_w"] = sum(iv["P_est"] for iv in intervals) / n
    out["frac_over_pcap"] = sum(1 for iv in intervals if iv["P_est"] > iv["Target"]) / n
    out["num_transitions"] = sum(
        1 for i in range(1, n)
        if abs(intervals[i]["f_max"] - intervals[i - 1]["f_max"]) > 0.001
    )
    out["avg_freq_ghz"] = sum(iv["f_max"] for iv in intervals) / n
    out["max_freq_ghz"] = max(iv["f_max"] for iv in intervals)

    return out


rows_metrics = []

for bench in sorted(os.listdir(TUNING_ROOT)):
    bench_dir = os.path.join(TUNING_ROOT, bench)
    if not os.path.isdir(bench_dir):
        continue

    n1_dir = os.path.join(bench_dir, "n1")
    if not os.path.isdir(n1_dir):
        continue

    for l3_name in sorted(os.listdir(n1_dir)):
        l3_dir = os.path.join(n1_dir, l3_name)
        if not os.path.isdir(l3_dir):
            continue

        m = re.match(r"l3_(\d+)MB$", l3_name)
        if not m:
            continue
        size_mb = int(m.group(1))

        sram7_dir = find_sram7_baseline(bench, size_mb)

        for variant in sorted(os.listdir(l3_dir)):
            run_dir = os.path.join(l3_dir, variant)
            if not os.path.isdir(run_dir):
                continue
            if not _has_results(run_dir):
                continue

            speedup = single_speedup_ratio(sram7_dir, run_dir)
            mets = metrics_from_run(run_dir)

            rows_metrics.append({
                "workload": bench,
                "size_mb": size_mb,
                "variant": variant,
                "run_dir": run_dir,
                "sram7_run_dir": sram7_dir,
                "speedup_vs_sram7": speedup,
                "mean_power_w": mets["mean_power_w"],
                "frac_over_pcap": mets["frac_over_pcap"],
                "num_transitions": mets["num_transitions"],
                "avg_freq_ghz": mets["avg_freq_ghz"],
                "max_freq_ghz": mets["max_freq_ghz"],
            })

df_metrics = pd.DataFrame(rows_metrics)
df_metrics = df_metrics.sort_values(["workload", "size_mb", "variant"]).reset_index(drop=True)

df_metrics_round = df_metrics.copy()
for c in ["speedup_vs_sram7", "mean_power_w", "frac_over_pcap", "avg_freq_ghz", "max_freq_ghz"]:
    df_metrics_round[c] = pd.to_numeric(df_metrics_round[c], errors="coerce").round(4)

df_metrics_round[
    [
        "workload",
        "size_mb",
        "variant",
        "speedup_vs_sram7",
        "mean_power_w",
        "frac_over_pcap",
        "num_transitions",
        "avg_freq_ghz",
        "max_freq_ghz",
    ]
]
```

```python
import numpy as np
import pandas as pd

# -----------------------------
# 1) One row per tuning setting
# -----------------------------
summary = df_metrics.copy()

summary["speedup_vs_sram7"] = pd.to_numeric(summary["speedup_vs_sram7"], errors="coerce")
summary["frac_over_pcap"] = pd.to_numeric(summary["frac_over_pcap"], errors="coerce")
summary["num_transitions"] = pd.to_numeric(summary["num_transitions"], errors="coerce")
summary["avg_freq_ghz"] = pd.to_numeric(summary["avg_freq_ghz"], errors="coerce")

def geomean_pos(xs):
    xs = pd.Series(xs).dropna()
    xs = xs[xs > 0]
    if len(xs) == 0:
        return np.nan
    return float(np.exp(np.log(xs).mean()))

variant_summary = (
    summary.groupby("variant", dropna=False)
    .agg(
        geomean_speedup=("speedup_vs_sram7", geomean_pos),
        mean_speedup=("speedup_vs_sram7", "mean"),
        mean_overcap=("frac_over_pcap", "mean"),
        worst_overcap=("frac_over_pcap", "max"),
        mean_transitions=("num_transitions", "mean"),
        mean_avg_freq=("avg_freq_ghz", "mean"),
        worst_speedup=("speedup_vs_sram7", "min"),
        n_workloads=("workload", "nunique"),
    )
    .reset_index()
)

for c in [
    "geomean_speedup",
    "mean_speedup",
    "mean_overcap",
    "worst_overcap",
    "mean_transitions",
    "mean_avg_freq",
    "worst_speedup",
]:
    variant_summary[c] = pd.to_numeric(variant_summary[c], errors="coerce").round(4)

variant_summary = variant_summary.sort_values(
    ["geomean_speedup", "mean_overcap"],
    ascending=[False, True]
).reset_index(drop=True)

print("\n=== Variant summary ===")
print(
    variant_summary[
        [
            "variant",
            "geomean_speedup",
            "mean_speedup",
            "mean_overcap",
            "worst_overcap",
            "mean_transitions",
            "mean_avg_freq",
            "worst_speedup",
            "n_workloads",
        ]
    ].to_string(index=False)
)

# -----------------------------------------
# 2) Per-workload winner under overcap cap
# -----------------------------------------
OVERCAP_LIMIT = 0.05

cand = df_metrics.copy()
cand["speedup_vs_sram7"] = pd.to_numeric(cand["speedup_vs_sram7"], errors="coerce")
cand["frac_over_pcap"] = pd.to_numeric(cand["frac_over_pcap"], errors="coerce")
cand["num_transitions"] = pd.to_numeric(cand["num_transitions"], errors="coerce")
cand["avg_freq_ghz"] = pd.to_numeric(cand["avg_freq_ghz"], errors="coerce")

winners = []

for workload, g in cand.groupby("workload"):
    g = g.dropna(subset=["speedup_vs_sram7"]).copy()
    if len(g) == 0:
        continue

    good = g[g["frac_over_pcap"] <= OVERCAP_LIMIT].copy()

    if len(good) > 0:
        row = good.sort_values(
            ["speedup_vs_sram7", "frac_over_pcap"],
            ascending=[False, True]
        ).iloc[0].copy()
        row["selection_mode"] = "within_limit"
    else:
        row = g.sort_values(
            ["frac_over_pcap", "speedup_vs_sram7"],
            ascending=[True, False]
        ).iloc[0].copy()
        row["selection_mode"] = "fallback_lowest_overcap"
    winners.append(row)

winner_table = pd.DataFrame(winners)[[
    "workload",
    "variant",
    "selection_mode",
    "speedup_vs_sram7",
    "frac_over_pcap",
    "num_transitions",
    "avg_freq_ghz",
]].sort_values("workload").reset_index(drop=True)

for c in ["speedup_vs_sram7", "frac_over_pcap", "avg_freq_ghz"]:
    winner_table[c] = pd.to_numeric(winner_table[c], errors="coerce").round(4)

print("\n=== Per-workload winner table ===")
print(winner_table.to_string(index=False))
```

```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import ListedColormap, BoundaryNorm
from matplotlib.lines import Line2D

# --------------------------------------------------
# Config
# --------------------------------------------------
CSV_PATH = "/home/skataoka26/COSC_498/devices/results/cache_sweep/uniform_370/nvsim_uniform_370.csv"
FREQ_GHZ = 2.2
F_W = 0.7

FW_VALUES = np.round(np.arange(0.0, 1.01, 0.1), 1)

# Heatmap settings
HEATMAP_CAPS = [16, 32, 128]
HEATMAP_EXCLUDE = []   # e.g. [("SRAM", 22)]

TARGET_COLORS = {
    "ReadLatency":  "#4C72B0",
    "WriteLatency": "#DD8452",
    "ReadEDP":      "#55A868",
    "WriteEDP":     "#C44E52",
}

# Keep raw nodes for computation, but display 14 as 16 everywhere
NODE_MARKERS = {7: "o", 14: "s", 22: "^", 32: "D"}
TECH_LINESTYLES = {"SRAM": "-", "MRAM": "--"}
TECH_LINE_COLORS = {"SRAM": "#2171B5", "MRAM": "#CB181D"}

# --------------------------------------------------
# Helpers
# --------------------------------------------------
def ns_to_cycles(ns, freq_ghz=FREQ_GHZ):
    return math.ceil(ns * freq_ghz)

def infer_tech(path):
    s = str(path)
    if "/SRAM/" in s:
        return "SRAM"
    if "/MRAM/" in s:
        return "MRAM"
    return "UNKNOWN"

def display_node_nm(node):
    """Show 14nm as 16nm in labels."""
    try:
        node = int(node)
    except Exception:
        return str(node)
    return 16 if node == 14 else node

def tech_node_label(tech, node):
    return f"{tech}  {display_node_nm(node)} nm"

def sweep_fw_tidy(df):
    """Return tidy DataFrame: one row per (tech, node, cap, fw) with chosen target."""
    rows = []
    for fw in FW_VALUES:
        tmp = df.copy()
        tmp["L_eff"] = (1 - fw) * tmp["lat_hit_ns"] + fw * tmp["lat_write_ns"]
        idx = tmp.groupby(["technology", "node_nm", "cap_mb"])["L_eff"].idxmin()
        opt = tmp.loc[idx]

        for _, r in opt.iterrows():
            rows.append({
                "technology": r["technology"],
                "node_nm": int(r["node_nm"]),
                "cap_mb": int(r["cap_mb"]),
                "fw": float(fw),
                "target": r["opt_target"],
            })
    return pd.DataFrame(rows)

def _pareto_front(points):
    """Return indices of Pareto-optimal points (minimise both x and y)."""
    pts = np.array(points)
    order = pts[:, 0].argsort()
    pts_sorted = pts[order]

    front = [order[0]]
    min_y = pts_sorted[0, 1]
    for i in range(1, len(pts_sorted)):
        if pts_sorted[i, 1] <= min_y:
            front.append(order[i])
            min_y = pts_sorted[i, 1]
    return sorted(front)

# --------------------------------------------------
# Load data
# --------------------------------------------------
df = pd.read_csv(CSV_PATH)
df["technology"] = df["stdout_path"].apply(infer_tech)

# Optional: keep only known techs
df = df[df["technology"].isin(["SRAM", "MRAM"])].copy()

# --------------------------------------------------
# Heatmap
# --------------------------------------------------
def build_heatmap(sweep_df):
    sweep_df = sweep_df[sweep_df["cap_mb"].isin(HEATMAP_CAPS)].copy()

    # Exclude SRAM 22nm from first plot
    sweep_df = sweep_df[
        ~((sweep_df["technology"] == "SRAM") & (sweep_df["node_nm"] == 22))
    ]

    for tech, node in HEATMAP_EXCLUDE:
        sweep_df = sweep_df[
            ~((sweep_df["technology"] == tech) & (sweep_df["node_nm"] == node))
        ]

    sweep_df["label"] = sweep_df.apply(
        lambda r: tech_node_label(r["technology"], r["node_nm"]), axis=1
    )

    # MRAM first, then SRAM; ascending node
    sweep_df["sort_key"] = sweep_df["technology"].map({"MRAM": 0, "SRAM": 1})
    sweep_df = sweep_df.sort_values(["cap_mb", "sort_key", "node_nm"])

    caps_ordered = sorted(sweep_df["cap_mb"].unique())

    label_order_df = (
        sweep_df[["label", "sort_key", "node_nm"]]
        .drop_duplicates()
        .sort_values(["sort_key", "node_nm"])
    )
    y_labels = label_order_df["label"].tolist()
    n_rows = len(y_labels)

    all_targets_seen = sorted(sweep_df["target"].dropna().unique())
    target_to_int = {t: i for i, t in enumerate(all_targets_seen)}
    n_targets = len(all_targets_seen)

    fw_list = list(np.round(FW_VALUES, 1))
    cap_matrices = {}

    for cap in caps_ordered:
        mat = np.full((n_rows, len(fw_list)), np.nan)
        group = sweep_df[sweep_df["cap_mb"] == cap]

        for ri, lbl in enumerate(y_labels):
            sub = group[group["label"] == lbl].sort_values("fw")
            for _, r in sub.iterrows():
                fw_val = round(float(r["fw"]), 1)
                if fw_val in fw_list and pd.notna(r["target"]):
                    ci = fw_list.index(fw_val)
                    mat[ri, ci] = target_to_int[r["target"]]

        cap_matrices[cap] = mat

    colors = [TARGET_COLORS.get(t, "#AAAAAA") for t in all_targets_seen]
    cmap = ListedColormap(colors)
    norm = BoundaryNorm(np.arange(-0.5, n_targets, 1), n_targets)

    n_caps = len(caps_ordered)
    fig, axes = plt.subplots(
        1,
        n_caps,
        figsize=(3.4 * n_caps, max(1.8, 0.28 * n_rows + 0.8)),
        sharey=True
    )
    if n_caps == 1:
        axes = [axes]

    for ai, cap in enumerate(caps_ordered):
        ax = axes[ai]
        ax.pcolormesh(
            cap_matrices[cap],
            cmap=cmap,
            norm=norm,
            edgecolors="black",
            linewidth=0.6
        )
        ax.set_title(f"{cap} MB", fontsize=10, fontweight="bold", pad=4)
        ax.set_xticks(np.arange(len(FW_VALUES)) + 0.5)
        ax.set_xticklabels([f"{v:.1f}" for v in FW_VALUES], fontsize=8)
        ax.set_xlabel(r"$f_w$", fontsize=10)
        ax.invert_yaxis()

    axes[0].set_yticks(np.arange(n_rows) + 0.5)
    axes[0].set_yticklabels(y_labels, fontsize=8, family="monospace")

    patches = [
        mpatches.Patch(facecolor=TARGET_COLORS.get(t, "#AAA"), edgecolor="black", label=t)
        for t in all_targets_seen
    ]
    fig.legend(
        handles=patches,
        title="Chosen Target",
        loc="upper center",
        bbox_to_anchor=(0.5, -0.05),
        ncol=max(1, len(patches)),
        fontsize=8,
        title_fontsize=9,
        frameon=True,
        edgecolor="black"
    )

    plt.tight_layout()
    plt.show()

# build + show
sweep_df = sweep_fw_tidy(df)
build_heatmap(sweep_df)

def build_pareto_scatter(df):
    """1x2 scatter of L_eff vs E_eff for 128 MB, split by technology."""
    sub = df[df["cap_mb"] == 128].copy()

    sub["L_eff"] = (1 - F_W) * sub["lat_hit_ns"] + F_W * sub["lat_write_ns"]
    sub["E_eff"] = (1 - F_W) * sub["dyn_energy_hit_nJ"] + F_W * sub["dyn_energy_write_nJ"]
    sub["node_int"] = sub["node_nm"].astype(int)

    fig, (ax_sram, ax_mram) = plt.subplots(1, 2, figsize=(12, 5), sharey=True)
    tech_axes = {"SRAM": ax_sram, "MRAM": ax_mram}

    tech_xlims = {}
    for tech in ["SRAM", "MRAM"]:
        t_sub = sub[sub["technology"] == tech]
        if t_sub.empty:
            tech_xlims[tech] = None
        else:
            xmin = t_sub["L_eff"].min()
            xmax = t_sub["L_eff"].max()
            pad = max((xmax - xmin) * 0.08, 0.5)
            tech_xlims[tech] = (xmin - pad, xmax + pad)

    # Force SRAM x-limit cutoff
    tech_xlims["SRAM"] = (0, 50)

    for tech, ax in tech_axes.items():
        t_sub = sub[sub["technology"] == tech].copy()
        if t_sub.empty:
            ax.set_title(tech, fontsize=12)
            ax.set_xlabel(r"$L_{\mathrm{eff}}$ (ns)", fontsize=11)
            ax.grid(True, linestyle="--", alpha=0.3)
            ax.spines["top"].set_visible(False)
            ax.spines["right"].set_visible(False)
            continue

        t_sub["node_disp"] = t_sub["node_int"].apply(display_node_nm)

        lat_rep_idx = set()
        eng_rep_idx = set()
        for node, grp in t_sub.groupby("node_disp"):
            lat_rep_idx.add(grp["L_eff"].idxmin())
            eng_rep_idx.add(grp["E_eff"].idxmin())
        rep_idx = lat_rep_idx | eng_rep_idx

        for idx, r in t_sub.iterrows():
            node_raw = int(r["node_int"])
            target = r["opt_target"]
            color = TARGET_COLORS.get(target, "#888888")
            marker = NODE_MARKERS.get(node_raw, "o")
            is_rep = idx in rep_idx

            ax.scatter(
                r["L_eff"], r["E_eff"],
                c=color,
                marker=marker,
                s=120 if is_rep else 60,
                edgecolors="black",
                linewidths=1.2 if is_rep else 0.5,
                alpha=1.0 if is_rep else 0.50,
                zorder=5 if is_rep else 3
            )

        if tech_xlims[tech] is not None:
            ax.set_xlim(*tech_xlims[tech])

        ax.set_title(tech, fontsize=12)
        ax.set_xlabel(r"$L_{\mathrm{eff}}$ (ns)", fontsize=11)
        ax.grid(True, linestyle="--", alpha=0.3)
        ax.spines["top"].set_visible(False)
        ax.spines["right"].set_visible(False)

    ax_sram.set_ylabel(r"$E_{\mathrm{eff}}$ (nJ)", fontsize=11)

    handles = []

    handles.append(Line2D([], [], linewidth=0, label="── Opt Target ──"))
    for tgt, clr in TARGET_COLORS.items():
        if tgt in sub["opt_target"].values:
            handles.append(
                mpatches.Patch(facecolor=clr, edgecolor="black", label=tgt)
            )

    handles.append(Line2D([], [], linewidth=0, label="── Node ──"))
    seen = set()
    for node in sorted(sub["node_int"].unique()):
        disp = display_node_nm(node)
        if disp in seen:
            continue
        seen.add(disp)
        mkr = NODE_MARKERS.get(int(node), "o")
        handles.append(
            Line2D([0], [0],
                   marker=mkr, color="w",
                   markerfacecolor="grey", markeredgecolor="black",
                   markersize=8, label=f"{disp} nm")
        )

    fig.legend(
        handles=handles,
        loc="upper center",
        bbox_to_anchor=(0.5, 1.08),
        ncol=len(handles),
        fontsize=8.5,
        frameon=False,
        handletextpad=0.4,
        columnspacing=1.0
    )

    plt.tight_layout()
    plt.show()

# show plot
build_pareto_scatter(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = ['2024', '2025', '2027', '2029', '2031', '2033', '2035', '2037', '2039']
vdd = [0.7, 0.65, 0.6, 0.6, 0.6, 0.6, 0.6, 0.6, 0.6]
power_density = [1.0, 1.08, 0.85, 0.97, 1.03, 1.09, 2.08, 3.55, 4.81]
logic_density = [19, 27, 34, 43, 57, 64, 153, 306, 459]
sram_density = [33, 40, 43, 45, 77, 77, 81, 81, 81]
target_freq = [3.10, 3.39, 3.51, 2.94, 4.17, 3.93, 4.18, 3.99, 3.82]
constrained_freq = [3.10, 3.14, 4.14, 3.02, 4.07, 3.60, 2.00, 1.13, 0.79]

# Setup Figure
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# --- Plot (a): Voltage vs Power Density ---
ax1 = axes[0]
color1 = 'tab:blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Supply Voltage ($V_{dd}$)', color=color1)
ax1.plot(years, vdd, marker='o', color=color1, linewidth=2, label='Vdd (V)')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.set_ylim(0.4, 0.8)
ax1.grid(True, linestyle='--', alpha=0.5)

ax2 = ax1.twinx()
color2 = 'tab:red'
ax2.set_ylabel('Normalized Power Density', color=color2)
ax2.plot(years, power_density, marker='s', color=color2, linewidth=2, linestyle='--', label='Power Density')
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylim(0, 5.5)
ax1.set_title('(a) The Power Wall')
ax1.set_xticks(years[::2]) # Reduce x-tick clutter

# --- Plot (b): Logic vs SRAM Density ---
ax3 = axes[1]
ax3.plot(years, logic_density, marker='^', color='tab:green', linewidth=2, label='Logic (Mgates/mm²)')
ax3.plot(years, sram_density, marker='d', color='tab:orange', linewidth=2, label='SRAM (Mbits/mm²)')
ax3.set_xlabel('Year')
ax3.set_ylabel('Density')
ax3.set_title('(b) The Density Divergence')
ax3.legend(loc='upper left')
ax3.grid(True, linestyle='--', alpha=0.5)
ax3.set_xticks(years[::2])

# --- Plot (c): Frequency Collapse ---
ax4 = axes[2]
ax4.plot(years, target_freq, marker='o', color='tab:purple', linewidth=2, label='Target Frequency')
ax4.plot(years, constrained_freq, marker='X', color='tab:brown', linewidth=2, linestyle='--', label='Thermally-Constrained')
ax4.set_xlabel('Year')
ax4.set_ylabel('CPU Frequency (GHz)')
ax4.set_title('(c) The Frequency Crisis')
ax4.legend(loc='lower left')
ax4.grid(True, linestyle='--', alpha=0.5)
ax4.set_ylim(0, 5)
ax4.set_xticks(years[::2])

fig.tight_layout()
plt.savefig('irds_combined_trends.png', dpi=300)
print("Combined plot generated successfully.")
```

```python
def plot_standalone_bar_crossnode(df_crossnode, sizes):
    _DESIGN_COLORS = {
        "SRAM 7nm":  "#4878CF",
        "SRAM 16nm": "#e8a838",
        "SRAM 32nm": "#D65F5F",
        "MRAM 16nm": "#6ACC64",
        "MRAM 32nm": "#B07AA1",
        "SRAM_7nm":  "#4878CF",
        "SRAM_16nm": "#e8a838",
        "SRAM_32nm": "#D65F5F",
        "MRAM_16nm": "#6ACC64",
        "MRAM_32nm": "#B07AA1",
    }

    # Hide these from plot/table
    HIDE_DESIGNS = {
        "SRAM 16nm", "SRAM_16nm",
        "SRAM 32nm", "SRAM_32nm",
        "MRAM 32nm", "MRAM_32nm",
    }

    ACTIVE_DESIGNS = [(n, t, v) for n, t, v in DESIGNS if n not in HIDE_DESIGNS]

    baseline_size = 16

    baseline_candidates = ["SRAM 7nm", "SRAM_7nm"]
    matches = []
    baseline_name = None
    for cand in baseline_candidates:
        matches = [(t, v) for n, t, v in ACTIVE_DESIGNS if n == cand]
        if matches:
            baseline_name = cand
            break

    assert matches, f"No SRAM 7nm baseline found in ACTIVE_DESIGNS: {ACTIVE_DESIGNS}"
    b_tech, b_var = matches[0]

    m_base = (
        (df_crossnode["size_mb"] == baseline_size) &
        (df_crossnode["tech"] == b_tech) &
        (df_crossnode["variant"].str.contains(b_var, na=False))
    )
    base_sub = df_crossnode[m_base].copy()
    assert not base_sub.empty, f"No baseline rows found for {baseline_name}, size={baseline_size}"

    base_benches = sorted(base_sub["bench"].dropna().unique())
    base_t = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "roi_time_s"])
        for b in base_benches
    }
    base_e = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "E_total_J"])
        for b in base_benches
    }

    benches = base_benches
    n_techs = len(ACTIVE_DESIGNS)

    x_positions = []
    x_labels = []
    sep_positions = []
    size_centers = []

    gap_between_sizes = 1.8
    cursor = 0.0

    for sz in sizes:
        group_x = []

        for b in benches:
            x_positions.append(cursor)
            x_labels.append(_short_bench(b))
            group_x.append(cursor)
            cursor += 1.0

        # dotted line before Mean
        sep_positions.append(cursor - 0.5)

        x_positions.append(cursor)
        x_labels.append("Mean")
        group_x.append(cursor)
        cursor += 1.0

        # dotted line after Mean
        sep_positions.append(cursor - 0.5)

        size_centers.append(np.mean(group_x))
        cursor += gap_between_sizes

    x_positions = np.array(x_positions, dtype=float)
    w = 0.8 / n_techs

    fig, (ax_top, ax_bot) = plt.subplots(
        2, 1,
        figsize=(max(14, 0.58 * len(x_positions)), 7.2),
        sharex=True
    )

    for ti, (dname, tech, variant) in enumerate(ACTIVE_DESIGNS):
        color = _DESIGN_COLORS.get(dname, "#999999")

        speedup_vals = []
        dyn_vals = []
        leak_vals = []

        for sz in sizes:
            m = (
                (df_crossnode["size_mb"] == sz) &
                (df_crossnode["tech"] == tech) &
                (df_crossnode["variant"].str.contains(variant, na=False))
            )
            sub = df_crossnode[m]

            per_bench_speedup = []
            per_bench_dyn = []
            per_bench_leak = []

            for b in benches:
                rows = sub[sub["bench"] == b]
                if rows.empty:
                    per_bench_speedup.append(np.nan)
                    per_bench_dyn.append(np.nan)
                    per_bench_leak.append(np.nan)
                    continue

                dt = np.mean(rows["roi_time_s"])
                de_dyn = np.mean(rows["E_dyn_J"])
                de_leak = np.mean(rows["E_leak_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)

                if any(np.isnan([bt, dt, be, de_dyn, de_leak])) or bt == 0 or be == 0:
                    per_bench_speedup.append(np.nan)
                    per_bench_dyn.append(np.nan)
                    per_bench_leak.append(np.nan)
                    continue

                per_bench_speedup.append(bt / dt)
                per_bench_dyn.append(de_dyn / be)
                per_bench_leak.append(de_leak / be)

            speedup_vals.extend(per_bench_speedup)
            dyn_vals.extend(per_bench_dyn)
            leak_vals.extend(per_bench_leak)

            valid_speed = [v for v in per_bench_speedup if not np.isnan(v)]
            valid_dyn = [v for v in per_bench_dyn if not np.isnan(v)]
            valid_leak = [v for v in per_bench_leak if not np.isnan(v)]

            speedup_vals.append(np.mean(valid_speed) if valid_speed else np.nan)
            dyn_vals.append(np.mean(valid_dyn) if valid_dyn else np.nan)
            leak_vals.append(np.mean(valid_leak) if valid_leak else np.nan)

        offset = (ti - n_techs / 2) * w + w / 2

        ax_top.bar(
            x_positions + offset,
            speedup_vals,
            width=w,
            color=color,
            edgecolor="black",
            linewidth=0.5,
            label=dname,
            zorder=3
        )

        ax_bot.bar(
            x_positions + offset,
            dyn_vals,
            width=w,
            color=color,
            alpha=0.9,
            edgecolor="black",
            linewidth=0.5,
            zorder=3
        )
        ax_bot.bar(
            x_positions + offset,
            leak_vals,
            width=w,
            bottom=np.nan_to_num(np.array(dyn_vals, dtype=float), nan=0.0),
            color=color,
            alpha=0.3,
            hatch='//',
            edgecolor="black",
            linewidth=0.5,
            zorder=3
        )

    for ax in (ax_top, ax_bot):
        ax.grid(axis='y', linestyle='--', alpha=0.3, zorder=0)
        ax.set_axisbelow(True)
        for s in sep_positions:
            ax.axvline(s, color="grey", linestyle=":", linewidth=1.2, alpha=0.9)

    ax_top.axhline(1.0, color="grey", ls="--", lw=1)
    ax_top.set_ylabel("Speedup", fontsize=11)

    ax_bot.axhline(1.0, color="grey", ls="--", lw=1)
    ax_bot.set_ylabel("Norm. L3 Energy", fontsize=11)

    ax_bot.set_xticks(x_positions)
    ax_bot.set_xticklabels(x_labels, rotation=45, ha="right", fontsize=12)

    ymin, ymax = ax_top.get_ylim()
    for xc, sz in zip(size_centers, sizes):
        ax_top.text(
            xc, ymax * 1.01, f"{sz}MB",
            ha="center", va="bottom",
            fontsize=11
        )

    design_handles, design_labels = ax_top.get_legend_handles_labels()
    fig.legend(
        design_handles, design_labels,
        loc="upper center",
        bbox_to_anchor=(0.5, 0.985),
        ncol=len(ACTIVE_DESIGNS),
        frameon=False,
        fontsize=10
    )

    dyn_patch = mpatches.Patch(
        facecolor='grey', alpha=0.9, edgecolor='black', label='Dynamic Energy'
    )
    leak_patch = mpatches.Patch(
        facecolor='grey', alpha=0.3, hatch='//', edgecolor='black', label='Leakage Energy'
    )
    ax_bot.legend(
        handles=[dyn_patch, leak_patch],
        loc="upper left",
        fontsize=9,
        framealpha=0.9
    )

    plt.tight_layout()
    plt.subplots_adjust(top=0.88, hspace=0.22)
    plt.show()


plot_standalone_bar_crossnode(df_cross, sizes=sizes)


def print_crossnode_text_table(df_crossnode, sizes):
    baseline_size = 16

    # Hide these from plot/table
    HIDE_DESIGNS = {
        "SRAM 16nm", "SRAM_16nm",
        "SRAM 32nm", "SRAM_32nm",
        "MRAM 32nm", "MRAM_32nm",
    }

    ACTIVE_DESIGNS = [(n, t, v) for n, t, v in DESIGNS if n not in HIDE_DESIGNS]

    baseline_candidates = ["SRAM 7nm", "SRAM_7nm"]
    matches = []
    baseline_name = None
    for cand in baseline_candidates:
        matches = [(t, v) for n, t, v in ACTIVE_DESIGNS if n == cand]
        if matches:
            baseline_name = cand
            break

    assert matches, f"No SRAM 7nm baseline found in ACTIVE_DESIGNS: {ACTIVE_DESIGNS}"
    b_tech, b_var = matches[0]

    m_base = (
        (df_crossnode["size_mb"] == baseline_size) &
        (df_crossnode["tech"] == b_tech) &
        (df_crossnode["variant"].str.contains(b_var, na=False))
    )
    base_sub = df_crossnode[m_base].copy()
    assert not base_sub.empty, f"No baseline rows found for {baseline_name}, size={baseline_size}"

    base_benches = sorted(base_sub["bench"].dropna().unique())
    base_t = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "roi_time_s"])
        for b in base_benches
    }
    base_e = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "E_total_J"])
        for b in base_benches
    }

    for sz in sizes:
        print(f"\n{'='*80}")
        print(f"SIZE: {sz}MB")
        print(f"{'='*80}")

        for dname, tech, variant in ACTIVE_DESIGNS:
            m = (
                (df_crossnode["size_mb"] == sz) &
                (df_crossnode["tech"] == tech) &
                (df_crossnode["variant"].str.contains(variant, na=False))
            )
            sub = df_crossnode[m]

            print(f"\n[{dname}]")
            print(f"{'bench':15s} {'speedup':>10s} {'dyn_norm':>10s} {'leak_norm':>10s} {'total_norm':>11s}")

            s_vals, d_vals, l_vals = [], [], []

            for b in base_benches:
                rows = sub[sub["bench"] == b]
                if rows.empty:
                    print(f"{_short_bench(b):15s} {'NA':>10s} {'NA':>10s} {'NA':>10s} {'NA':>11s}")
                    continue

                dt = np.mean(rows["roi_time_s"])
                de_dyn = np.mean(rows["E_dyn_J"])
                de_leak = np.mean(rows["E_leak_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)

                if any(np.isnan([bt, dt, be, de_dyn, de_leak])) or bt == 0 or be == 0:
                    print(f"{_short_bench(b):15s} {'NA':>10s} {'NA':>10s} {'NA':>10s} {'NA':>11s}")
                    continue

                speedup = bt / dt
                dyn_norm = de_dyn / be
                leak_norm = de_leak / be
                total_norm = dyn_norm + leak_norm

                s_vals.append(speedup)
                d_vals.append(dyn_norm)
                l_vals.append(leak_norm)

                print(f"{_short_bench(b):15s} {speedup:10.4f} {dyn_norm:10.4f} {leak_norm:10.4f} {total_norm:11.4f}")

            if s_vals:
                print(f"{'-'*80}")
                print(
                    f"{'Mean':15s} "
                    f"{np.mean(s_vals):10.4f} "
                    f"{np.mean(d_vals):10.4f} "
                    f"{np.mean(l_vals):10.4f} "
                    f"{(np.mean(d_vals) + np.mean(l_vals)):11.4f}"
                )


print_crossnode_text_table(df_cross, sizes)
```

```python
def plot_standalone_scatter_hca(df_hybrid, sizes, save_path=None, print_summary=True):
    hca_baseline_tech = "sram14_mram14"
    hca_baseline_var = "baseline_mram_only"
    hca_baseline_label = "All-MRAM 16nm"

    hca_benches = {}
    for sz in sizes:
        bench_sets = []

        m = (
            (df_hybrid["size_mb"] == sz) &
            (df_hybrid["tech"] == hca_baseline_tech) &
            (df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        )
        bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())

        for _, tech, var in _STATIC_SCENARIOS:
            m = (
                (df_hybrid["size_mb"] == sz) &
                (df_hybrid["tech"] == tech) &
                (df_hybrid["variant"].str.contains(var, na=False))
            )
            bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())

        valid = [s for s in bench_sets if s]
        hca_benches[sz] = sorted(set().union(*valid)) if valid else []

    all_hca_benches = sorted(set().union(*(hca_benches.values())))
    hca_bench_marker = {
        b: _BENCH_MARKERS[i % len(_BENCH_MARKERS)]
        for i, b in enumerate(all_hca_benches)
    }

    summary_rows = []

    # horizontal layout: 1 row, N columns
    ncols = len(sizes)
    fig, axes = plt.subplots(1, ncols, figsize=(4.8 * ncols, 4.2))
    if ncols == 1:
        axes = [axes]

    for ci, sz in enumerate(sizes):
        ax = axes[ci]
        benches = hca_benches.get(sz, [])

        m = (
            (df_hybrid["size_mb"] == sz) &
            (df_hybrid["tech"] == hca_baseline_tech) &
            (df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        )
        base_sub = df_hybrid[m]

        print(f"\n[debug] size={sz} baseline rows={len(base_sub)}")
        if len(base_sub):
            print(
                base_sub[["bench", "tech", "variant", "roi_time_s", "E_total_J", "times_s"]]
                .head()
                .to_string(index=False)
            )

        base_t = {}
        base_e = {}
        base_row_map = {}

        for b in benches:
            rows_b = base_sub[base_sub["bench"] == b]
            if rows_b.empty:
                continue
            base_t[b] = gmean(rows_b["roi_time_s"])
            base_e[b] = gmean(rows_b["E_total_J"])
            base_row_map[b] = rows_b.iloc[0]

        for si, (label, tech, var) in enumerate(_STATIC_SCENARIOS):
            m = (
                (df_hybrid["size_mb"] == sz) &
                (df_hybrid["tech"] == tech) &
                (df_hybrid["variant"].str.contains(var, na=False))
            )
            sub = df_hybrid[m]
            color = _COMBINED_COLORS.get(label, "#999")

            xs_pts, ys_pts = [], []
            for b in benches:
                rows = sub[sub["bench"] == b]
                if rows.empty or b not in base_row_map:
                    continue

                dt = gmean(rows["roi_time_s"])
                de = gmean(rows["E_total_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)

                if np.isnan(bt) or np.isnan(dt) or np.isnan(be) or np.isnan(de):
                    continue
                if bt <= 0 or dt <= 0 or be <= 0 or de <= 0:
                    continue

                base_row = base_row_map[b]
                cand_row = rows.iloc[0]
                ws_over_n, ws_terms, active_idx = _wsn_active_from_rows(base_row, cand_row)
                norm_energy = de / be

                if not np.isfinite(ws_over_n):
                    print(f"[warn] missing active-core WS/N inputs: size={sz} bench={b} scenario={label}")
                    continue

                summary_rows.append({
                    "size_mb": sz,
                    "bench": b,
                    "scenario": label,
                    "ws_over_n": ws_over_n,
                    "norm_energy": norm_energy,
                    "base_time_s": bt,
                    "cand_time_s": dt,
                    "base_energy_J": be,
                    "cand_energy_J": de,
                    "n_terms": len(ws_terms),
                    "active_core_idx": ",".join(str(i) for i in active_idx),
                    "ws_terms": "|".join(f"{x:.9f}" for x in ws_terms),
                })

                ax.scatter(
                    ws_over_n, norm_energy,
                    marker=hca_bench_marker[b],
                    s=60, alpha=0.95, zorder=3 + si,
                    facecolors=color, edgecolors="#444444", linewidths=1.2
                )
                xs_pts.append(ws_over_n)
                ys_pts.append(norm_energy)

            if xs_pts:
                ax.scatter(
                    [gmean(xs_pts)], [gmean(ys_pts)],
                    marker="D", s=180, alpha=1.0,
                    facecolors=color, edgecolors="black",
                    linewidths=1.5, zorder=7 + si
                )

        ax.axhline(1.0, color="grey", ls="--", lw=1)
        ax.axvline(1.0, color="grey", ls="--", lw=1)
        ax.set_xlim(0.8, 1.1)
        ax.set_ylim(bottom=0)

        ax.set_title(f"Static {sz}MB", fontsize=11)
        ax.set_xlabel("Speedup", fontsize=10)

        if ci == 0:
            ax.set_ylabel("Norm. L3 Energy", fontsize=11)

    hca_handles = [
        Line2D(
            [0], [0], marker="D", color="w",
            markerfacecolor=_COMBINED_COLORS.get(lab, "#999"),
            markersize=8, markeredgecolor="k", markeredgewidth=1.0, label=lab
        )
        for lab in ["All-MRAM 16nm", "All-SRAM 16nm", "Static 4 SRAM", "Static 8 SRAM", "Static 12 SRAM"]
    ]

    bench_handles = [
        Line2D(
            [0], [0], marker=hca_bench_marker[b], color="w",
            markerfacecolor="grey", markersize=6,
            markeredgecolor="k", markeredgewidth=0.5,
            label=_short_bench(b)
        )
        for b in all_hca_benches
    ]

    plt.tight_layout()
    plt.subplots_adjust(top=0.80)

    fig.legend(
        handles=hca_handles + bench_handles,
        loc="upper center",
        bbox_to_anchor=(0.5, 1.02),
        ncol=5,
        frameon=False,
        fontsize=8
    )

    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")

    summary_df = pd.DataFrame(summary_rows)
    if print_summary:
        _print_summary_tables(summary_df)

    plt.show()
    return summary_df

summary_df = plot_standalone_scatter_hca(df_hybrid, sizes=sizes, print_summary=True)
```

```python
import re
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import yaml

BASE_DIR = Path("~/COSC_498/miniMXE").expanduser()
ALL_CSV = Path("~/COSC_498/miniMXE/repro/agg/hca.csv").expanduser()

ASSOC = 16
TARGET_SIZE_MB = 16

DEVICE_DIR = BASE_DIR / "mx3" / "config" / "devices" / "sunnycove"


def load_nvsim_params(device_dir):
    def _read_yaml(fname):
        with open(device_dir / fname, "r") as f:
            raw = yaml.safe_load(f) or {}
        out = {}
        for size_mb, vals in raw.items():
            out[str(size_mb)] = {
                "r_pj": float(vals.get("r_pj", 0.0)),
                "w_pj": float(vals.get("w_pj", 0.0)),
                "leak_mw": float(vals.get("leak_mw", 0.0)),
            }
        return out

    sram7 = _read_yaml("sram7.yaml")
    sram14 = _read_yaml("sram14.yaml")
    sram32 = _read_yaml("sram32.yaml")
    mram14 = _read_yaml("mram14.yaml")
    mram32 = _read_yaml("mram32.yaml")

    params = {}

    params["sram7"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram7.items()
    }

    params["sram14"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram14.items()
    }

    params["sram32"] = {
        sz: {
            "SRAM_R_pJ": v["r_pj"],
            "SRAM_W_pJ": v["w_pj"],
            "SRAM_LEAK_mW": v["leak_mw"],
            "MRAM_R_pJ": 0.0,
            "MRAM_W_pJ": 0.0,
            "MRAM_LEAK_mW": 0.0,
        }
        for sz, v in sram32.items()
    }

    params["mram14"] = {}
    for sz in sorted(set(sram14.keys()) | set(mram14.keys()), key=lambda x: int(x)):
        sv = sram14.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        mv = mram14.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        params["mram14"][sz] = {
            "SRAM_R_pJ": sv["r_pj"],
            "SRAM_W_pJ": sv["w_pj"],
            "SRAM_LEAK_mW": sv["leak_mw"],
            "MRAM_R_pJ": mv["r_pj"],
            "MRAM_W_pJ": mv["w_pj"],
            "MRAM_LEAK_mW": mv["leak_mw"],
        }

    params["mram32"] = {}
    for sz in sorted(set(sram32.keys()) | set(mram32.keys()), key=lambda x: int(x)):
        sv = sram32.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        mv = mram32.get(sz, {"r_pj": 0.0, "w_pj": 0.0, "leak_mw": 0.0})
        params["mram32"][sz] = {
            "SRAM_R_pJ": sv["r_pj"],
            "SRAM_W_pJ": sv["w_pj"],
            "SRAM_LEAK_mW": sv["leak_mw"],
            "MRAM_R_pJ": mv["r_pj"],
            "MRAM_W_pJ": mv["w_pj"],
            "MRAM_LEAK_mW": mv["leak_mw"],
        }

    params["sram14_mram14"] = dict(params["mram14"])
    return params


NVSIM_PARAMS = load_nvsim_params(DEVICE_DIR)

_STATIC_SCENARIOS = [
    ("All-MRAM 16nm", "mram14", "baseline_mram_only"),
    ("All-SRAM 16nm", "sram14", "baseline_sram_only"),
    ("Static 4 SRAM", "sram14_mram14", "noparity_s4_fillmram_sram14_mram14"),
    ("Static 8 SRAM", "sram14_mram14", "noparity_s8_fillmram_sram14_mram14"),
    ("Static 12 SRAM", "sram14_mram14", "noparity_s12_fillmram_sram14_mram14"),
]

_COMBINED_COLORS = {
    "All-MRAM 16nm": "#B22222",
    "All-SRAM 16nm": "#999999",
    "Static 4 SRAM": "#1f77b4",
    "Static 8 SRAM": "#2ca02c",
    "Static 12 SRAM": "#17becf",
}

_BENCH_MARKERS = ["o", "v", "^", "<", ">", "s", "p", "*", "h", "H", "D", "d", "P", "X"]

plt.rcParams.update({
    "font.size": 12,
    "axes.titlesize": 13,
    "axes.labelsize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 9,
    "figure.dpi": 150,
    "savefig.dpi": 150,
    "savefig.bbox": "tight",
})


def gmean(xs):
    xs = np.asarray(list(xs), dtype=float)
    xs = xs[np.isfinite(xs) & (xs > 0)]
    return float(np.exp(np.mean(np.log(xs)))) if xs.size else float("nan")


def _short_bench(name):
    parts = name.split("+")
    shortened = []
    for p in parts:
        s = re.sub(r"^\d+_", "", p)
        s = re.sub(r"_(r|s)$", "", s)
        shortened.append(s)
    return "+".join(shortened)


def compute_w_sram(df_in):
    v = df_in["variant"].fillna("").astype(str)
    w = np.full(len(df_in), np.nan, dtype=float)

    w[v.str.contains("baseline_sram_only", na=False).values] = 1.0
    w[v.str.contains("baseline_mram_only", na=False).values] = 0.0

    sw = pd.to_numeric(df_in.get("sram_ways", np.nan), errors="coerce").values
    ft = df_in.get("fill_to", "").fillna("").astype(str).str.lower().values

    for i in range(len(w)):
        if np.isfinite(w[i]):
            continue
        try:
            frac = float(sw[i]) / ASSOC
        except Exception:
            frac = 0.5
        if ft[i] == "sram":
            frac = max(frac, 0.6)
        if ft[i] == "mram":
            frac = min(frac, 0.4)
        w[i] = np.clip(frac, 0.0, 1.0)

    return w


def load_all():
    print(f"Reading {ALL_CSV}...")
    df = pd.read_csv(ALL_CSV)
    print(f"Read {len(df)} rows.")

    df.rename(columns={
        "workload": "bench",
        "device": "tech",
        "instructions": "insn",
        "l3_rh_sram": "read_hits_sram",
        "l3_rh_mram": "read_hits_mram",
        "l3_wh_sram": "write_hits_sram",
        "l3_wh_mram": "write_hits_mram",
    }, inplace=True)

    df["size_mb"] = (
        df["capacity"].astype(str)
        .str.replace(r"^sz", "", regex=True)
        .str.replace(r"M$", "", regex=True)
    )
    df["size_mb"] = pd.to_numeric(df["size_mb"], errors="coerce")
    df["roi_time_s"] = df["elapsed_ns"] / 1e9

    def _build_rw_tag(row):
        ls = row.get("lat_scale")
        dev = str(row.get("tech", ""))
        if pd.isna(ls) or str(ls).strip() == "":
            m_dev = re.search(r"_rd(\d+)x", dev)
            if m_dev:
                return f"r{m_dev.group(1)}x"
            return ""
        m = re.search(r"(r\d+x_w\d+x)", dev)
        if m:
            return m.group(1)
        m2 = re.search(r"rd(\d+)x", dev)
        if m2:
            return f"r{m2.group(1)}x"
        return str(ls)

    df["rw_tag"] = df.apply(_build_rw_tag, axis=1)

    df["tech"] = (
        df["tech"].astype(str)
        .str.replace(r"_r\d+x_w\d+x$", "", regex=True)
        .str.replace(r"_rd\d+x$", "", regex=True)
    )

    def _synth_variant(row):
        hca = str(row.get("hca", "")).strip().lower()
        tech = str(row.get("tech", ""))
        wp = str(row.get("way_policy", "")).strip().lower() if pd.notna(row.get("way_policy")) else ""
        sm = str(row.get("static_or_migration", "")).strip().lower() if pd.notna(row.get("static_or_migration")) else ""
        rf = str(row.get("restricted_fill", "")).strip().lower() if pd.notna(row.get("restricted_fill")) else ""
        s_val = row.get("s")
        p_val = row.get("p")
        c_val = row.get("c")

        if hca != "yes":
            if "mram" in tech and "sram" not in tech:
                return "baseline_mram_only"
            elif "sram" in tech and "mram" not in tech:
                return "baseline_sram_only"
            elif "mram" in tech:
                return "baseline_mram_only"
            return "baseline"

        parts = []
        if wp:
            parts.append(wp)
        if pd.notna(s_val) and str(s_val).strip() != "":
            parts.append(f"s{int(float(s_val))}")
        parts.append("fillmram")
        if rf == "yes":
            parts.append("rf")
        if sm == "migration":
            if pd.notna(p_val) and str(p_val).strip() != "":
                parts.append(f"p{int(float(p_val))}")
            if pd.notna(c_val) and str(c_val).strip() != "":
                parts.append(f"c{int(float(c_val))}")
        parts.append(tech)
        return "_".join(parts)

    df["variant"] = df.apply(_synth_variant, axis=1)

    df["read_hits"] = df["read_hits_sram"].fillna(0) + df["read_hits_mram"].fillna(0)
    df["write_hits"] = df["write_hits_sram"].fillna(0) + df["write_hits_mram"].fillna(0)
    df["misses"] = df["dram_reads"].fillna(0) + df["dram_writes"].fillna(0)

    def _sram_ways(row):
        v = str(row["variant"])
        if "baseline_sram_only" in v:
            return ASSOC
        if "baseline_mram_only" in v:
            return 0
        s = row.get("s")
        if pd.notna(s) and str(s).strip() != "":
            try:
                return int(float(s))
            except Exception:
                pass
        return 0

    df["sram_ways"] = df.apply(_sram_ways, axis=1).astype(int)
    df["fill_to"] = df["variant"].apply(
        lambda v: "sram" if "baseline_sram_only" in v else "mram"
    )
    df["fill_to_l"] = df["fill_to"].str.lower()

    nums = [
        "roi_time_s", "read_hits", "write_hits", "misses", "size_mb",
        "sram_ways", "read_hits_sram", "read_hits_mram",
        "write_hits_sram", "write_hits_mram", "insn"
    ]
    for c in nums:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    df["variant"] = df["variant"].astype(str)
    df["tech"] = df["tech"].astype(str)
    df["bench"] = df["bench"].astype(str)

    ok = df["roi_time_s"].fillna(0) > 0
    exclude_mask = (
        df["bench"].str.contains("lbm", case=False) |
        df["bench"].str.contains("619", case=False) |
        df["bench"].str.contains("wrf", case=False) |
        df["bench"].str.contains("621", case=False)
    )
    df = df[ok & (~exclude_mask)].copy()
    print(f"After filtering: {len(df)} rows")
    return df


def calc_energy(df):
    df = df.copy()
    df["w_sram"] = compute_w_sram(df)

    dyn_J = np.full(len(df), np.nan)
    leak_J = np.full(len(df), np.nan)

    nvsim_tech_map = {
        "sram14": "mram14",
        "sram14_mram14": "mram14",
    }

    for i in range(len(df)):
        tech = df.iloc[i]["tech"]
        lookup_tech = nvsim_tech_map.get(tech, tech)
        size_mb = int(df.iloc[i]["size_mb"]) if pd.notna(df.iloc[i]["size_mb"]) else None
        if size_mb is None:
            continue

        p = NVSIM_PARAMS.get(lookup_tech, {}).get(str(size_mb))
        if p is None:
            continue

        frac = df["w_sram"].iloc[i]
        rh_s = df.iloc[i].get("read_hits_sram", 0) or 0
        rh_m = df.iloc[i].get("read_hits_mram", 0) or 0
        wh_s = df.iloc[i].get("write_hits_sram", 0) or 0
        wh_m = df.iloc[i].get("write_hits_mram", 0) or 0

        if (rh_s + rh_m + wh_s + wh_m) > 0:
            dyn_pJ = (
                rh_s * p["SRAM_R_pJ"] + rh_m * p.get("MRAM_R_pJ", 0.0) +
                wh_s * p["SRAM_W_pJ"] + wh_m * p.get("MRAM_W_pJ", 0.0)
            )
        else:
            r = df.iloc[i].get("read_hits", 0) or 0
            w = df.iloc[i].get("write_hits", 0) or 0
            dyn_pJ = (
                r * (frac * p["SRAM_R_pJ"] + (1 - frac) * p.get("MRAM_R_pJ", 0.0)) +
                w * (frac * p["SRAM_W_pJ"] + (1 - frac) * p.get("MRAM_W_pJ", 0.0))
            )

        misses = df.iloc[i].get("misses", 0) or 0
        fill_to = df.iloc[i].get("fill_to_l", "mram")
        fill_w = p["SRAM_W_pJ"] if fill_to == "sram" else p.get("MRAM_W_pJ", 0.0)
        dyn_pJ += misses * fill_w
        dyn_J[i] = dyn_pJ * 1e-12

        leak_mW = frac * p["SRAM_LEAK_mW"] + (1 - frac) * p.get("MRAM_LEAK_mW", 0.0)
        t = df.iloc[i]["roi_time_s"]
        if pd.notna(t) and t > 0:
            leak_J[i] = (leak_mW * 1e-3) * t

    df["E_dyn_J"] = dyn_J
    df["E_leak_J"] = leak_J
    df["E_total_J"] = dyn_J + leak_J
    return df


def plot_hybrid_only_16mb_square(df_hybrid, save_path=None):
    baseline_tech = "mram14"
    baseline_var = "baseline_mram_only"
    baseline_label = "All-MRAM 16nm"
    size_mb = TARGET_SIZE_MB

    bench_sets = []
    for _, tech, var in _STATIC_SCENARIOS:
        m = (
            (df_hybrid["size_mb"] == size_mb) &
            (df_hybrid["tech"] == tech) &
            (df_hybrid["variant"].str.contains(var, na=False))
        )
        bench_sets.append(set(df_hybrid.loc[m, "bench"].unique()))

    valid_sets = [s for s in bench_sets if s]
    benches = sorted(set().union(*valid_sets)) if valid_sets else []
    if not benches:
        print(f"No hybrid data found for {size_mb}MB.")
        return

    all_bench_markers = {
        b: _BENCH_MARKERS[i % len(_BENCH_MARKERS)]
        for i, b in enumerate(benches)
    }

    m_base = (
        (df_hybrid["size_mb"] == size_mb) &
        (df_hybrid["tech"] == baseline_tech) &
        (df_hybrid["variant"].str.contains(baseline_var, na=False))
    )
    base_sub = df_hybrid[m_base].copy()

    base_t = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "roi_time_s"])
        for b in benches
        if not base_sub.loc[base_sub["bench"] == b].empty
    }
    base_e = {
        b: gmean(base_sub.loc[base_sub["bench"] == b, "E_total_J"])
        for b in benches
        if not base_sub.loc[base_sub["bench"] == b].empty
    }

    fig, ax = plt.subplots(figsize=(4.5, 4.5))  # square

    for si, (label, tech, var) in enumerate(_STATIC_SCENARIOS):
        m = (
            (df_hybrid["size_mb"] == size_mb) &
            (df_hybrid["tech"] == tech) &
            (df_hybrid["variant"].str.contains(var, na=False))
        )
        sub = df_hybrid[m].copy()
        color = _COMBINED_COLORS.get(label, "#999999")

        xs_pts, ys_pts = [], []
        for b in benches:
            rows = sub[sub["bench"] == b]
            if rows.empty:
                continue

            dt = gmean(rows["roi_time_s"])
            de = gmean(rows["E_total_J"])
            bt = base_t.get(b, np.nan)
            be = base_e.get(b, np.nan)

            if any(np.isnan([bt, dt, be, de])) or bt == 0 or be == 0:
                continue

            x = bt / dt
            y = de / be

            ax.scatter(
                x, y,
                marker=all_bench_markers[b],
                s=70,
                alpha=0.95,
                zorder=3 + si,
                facecolors=color,
                edgecolors="#444444",
                linewidths=1.2,
            )
            xs_pts.append(x)
            ys_pts.append(y)

        if xs_pts:
            ax.scatter(
                [np.mean(xs_pts)],
                [np.mean(ys_pts)],
                marker="D",
                s=190,
                alpha=1.0,
                facecolors=color,
                edgecolors="black",
                linewidths=1.5,
                zorder=10 + si,
            )

    ax.axhline(1.0, color="grey", ls="--", lw=1)
    ax.axvline(1.0, color="grey", ls="--", lw=1)
    ax.set_ylim(bottom=0)
    ax.set_xlabel(f"Speedup vs {baseline_label}", fontsize=11)
    ax.set_ylabel("Norm. L3 Energy", fontsize=11)
    ax.set_title("Static SRAM/MRAM Partitioning (16MB)", fontsize=12)

    ax.text(
        0.03, 0.04, "16MB",
        transform=ax.transAxes,
        ha="left", va="bottom", fontsize=11,
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.8, edgecolor="gray"),
    )

    design_handles = [
        Line2D(
            [0], [0],
            marker="D",
            color="w",
            markerfacecolor=_COMBINED_COLORS.get(lab, "#999999"),
            markersize=8,
            markeredgecolor="k",
            markeredgewidth=1.0,
            label=lab,
        )
        for lab in [
            "All-MRAM 16nm",
            "All-SRAM 16nm",
            "Static 4 SRAM",
            "Static 8 SRAM",
            "Static 12 SRAM",
        ]
    ]

    bench_handles = [
        Line2D(
            [0], [0],
            marker=all_bench_markers[b],
            color="w",
            markerfacecolor="grey",
            markersize=6,
            markeredgecolor="k",
            markeredgewidth=0.5,
            label=_short_bench(b),
        )
        for b in benches
    ]

    plt.tight_layout()
    plt.subplots_adjust(top=0.78)
    fig.legend(
        handles=design_handles + bench_handles,
        loc="upper center",
        bbox_to_anchor=(0.5, 1.02),
        ncol=5,
        frameon=False,
        fontsize=8,
    )

    if save_path:
        fig.savefig(save_path, dpi=150, bbox_inches="tight")
        print(f"Saved: {save_path}")

    plt.show()


if __name__ == "__main__":
    df_all = load_all()
    print(f"\nAll techs: {sorted(df_all['tech'].unique())}")
    print(f"All sizes: {sorted(df_all['size_mb'].dropna().unique())}")

    is_baseline = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "no") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_cross = calc_energy(df_all[is_baseline].copy())
    print(f"\nCrossnode: {len(df_cross)} rows")

    is_hca = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "yes") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_hca = calc_energy(df_all[is_hca].copy())
    print(f"HCA: {len(df_hca)} rows")

    df_baselines = df_cross[
        df_cross["tech"].isin(["mram14", "sram14"]) &
        (
            df_cross["variant"].str.contains("baseline_mram_only", na=False) |
            df_cross["variant"].str.contains("baseline_sram_only", na=False)
        )
    ].copy()

    df_hybrid = pd.concat([df_hca, df_baselines], ignore_index=True)
    df_hybrid = df_hybrid[df_hybrid["size_mb"] == TARGET_SIZE_MB].copy()
    print(f"Hybrid {TARGET_SIZE_MB}MB: {len(df_hybrid)} rows")

    plot_hybrid_only_16mb_square(
        df_hybrid,
        save_path=BASE_DIR / f"hybrid_only_{TARGET_SIZE_MB}MB_square.png",
    )

    print("\nDone!")
```

```python
def print_hybrid_scatter_points_and_means(df_hybrid, sizes=(16, 32, 128)):
    hca_baseline_tech = "mram14"
    hca_baseline_var = "baseline_mram_only"
    hca_baseline_label = "All-MRAM 16nm"

    for sz in sizes:
        print("\n" + "=" * 80)
        print(f"HYBRID SCATTER DATA FOR {sz}MB  (baseline = {hca_baseline_label})")
        print("=" * 80)

        bench_sets = []
        m = (
            (df_hybrid["size_mb"] == sz) &
            (df_hybrid["tech"] == hca_baseline_tech) &
            (df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        )
        bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())

        for _, tech, var in _STATIC_SCENARIOS:
            m = (
                (df_hybrid["size_mb"] == sz) &
                (df_hybrid["tech"] == tech) &
                (df_hybrid["variant"].str.contains(var, na=False))
            )
            bench_sets.append(set(df_hybrid[m]["bench"].unique()) if m.any() else set())

        valid = [s for s in bench_sets if s]
        benches = sorted(set().union(*valid)) if valid else []

        if not benches:
            print("No data found.")
            continue

        m = (
            (df_hybrid["size_mb"] == sz) &
            (df_hybrid["tech"] == hca_baseline_tech) &
            (df_hybrid["variant"].str.contains(hca_baseline_var, na=False))
        )
        base_sub = df_hybrid[m].copy()

        base_t = {
            b: gmean(base_sub[base_sub["bench"] == b]["roi_time_s"])
            for b in benches
            if not base_sub[base_sub["bench"] == b].empty
        }
        base_e = {
            b: gmean(base_sub[base_sub["bench"] == b]["E_total_J"])
            for b in benches
            if not base_sub[base_sub["bench"] == b].empty
        }

        for label, tech, var in _STATIC_SCENARIOS:
            print(f"\n--- {label} ---")
            m = (
                (df_hybrid["size_mb"] == sz) &
                (df_hybrid["tech"] == tech) &
                (df_hybrid["variant"].str.contains(var, na=False))
            )
            sub = df_hybrid[m].copy()

            xs_pts, ys_pts = [], []

            for b in benches:
                rows = sub[sub["bench"] == b]
                if rows.empty:
                    continue

                dt = gmean(rows["roi_time_s"])
                de = gmean(rows["E_total_J"])
                bt = base_t.get(b, np.nan)
                be = base_e.get(b, np.nan)

                if np.isnan(bt) or np.isnan(dt) or np.isnan(be) or np.isnan(de):
                    continue
                if bt == 0 or be == 0:
                    continue

                speedup = bt / dt
                norm_energy = de / be

                xs_pts.append(speedup)
                ys_pts.append(norm_energy)

                print(
                    f"{_short_bench(b):20s}  "
                    f"x(speedup) = {speedup:8.4f}   "
                    f"y(norm_energy) = {norm_energy:8.4f}"
                )

            if xs_pts:
                mean_speedup = float(np.mean(xs_pts))
                mean_energy = float(np.mean(ys_pts))
                print(f"MEAN -> speedup = {mean_speedup:.4f}, norm_energy = {mean_energy:.4f}")
            else:
                print("No valid points.")


# call it after df_hybrid is built
if __name__ == "__main__":
    df_all = load_all()

    # Crossnode slice
    is_baseline = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "no") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_cross = calc_energy(df_all[is_baseline].copy())

    # HCA slice
    is_hca = (
        (df_all["hca"].fillna("").astype(str).str.lower() == "yes") &
        (df_all["rw_tag"].fillna("") == "")
    )
    df_hca = calc_energy(df_all[is_hca].copy())

    # Baselines to merge with HCA for hybrid analysis
    df_baselines = df_cross[
        df_cross["tech"].isin(["mram14", "sram14"]) &
        (
            df_cross["variant"].str.contains("baseline_mram_only", na=False) |
            df_cross["variant"].str.contains("baseline_sram_only", na=False)
        )
    ].copy()

    # IMPORTANT: keep the FULL hybrid dataframe
    df_hybrid = pd.concat([df_hca, df_baselines], ignore_index=True)

    print(f"Hybrid rows total: {len(df_hybrid)}")
    print("Hybrid sizes present:", sorted(df_hybrid["size_mb"].dropna().astype(int).unique()))

    # print points for 16/32/128
    print_hybrid_scatter_points_and_means(df_hybrid, sizes=(16, 32, 128))

    # if you still want the square 16MB-only plot, filter ONLY here
    plot_hybrid_only_16mb_square(
        df_hybrid[df_hybrid["size_mb"] == 16].copy(),
        save_path=BASE_DIR / "hybrid_only_16MB_square.png",
    )
```
